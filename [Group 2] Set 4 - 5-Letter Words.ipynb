# 46 (L) GTCAP - (C) AGI - (E) ALI - (ε) URC - (W) PGOLD

```python
#extract the overlapping dates

start_46 = max(min(Date_GTCAP), min(Date_AGI), min(Date_ALI), min(Date_URC), min(Date_PGOLD))
end_46 = min(max(Date_GTCAP), max(Date_AGI), max(Date_ALI), max(Date_URC), max(Date_PGOLD))

#shorten the closing prices
GTCAP_46_1 = Date_GTCAP.index(start_46)
GTCAP_46_2 = Date_GTCAP.index(end_46)

AGI_46_1 = Date_AGI.index(start_46)
AGI_46_2 = Date_AGI.index(end_46)

ALI_46_1 = Date_ALI.index(start_46)
ALI_46_2 = Date_ALI.index(end_46)

URC_46_1 = Date_URC.index(start_46)
URC_46_2 = Date_URC.index(end_46)

PGOLD_46_1 = Date_PGOLD.index(start_46)
PGOLD_46_2 = Date_PGOLD.index(end_46)

#save the shortened versions of closing prices
L46 = np.save('GTCAPshortPrices.npy', ClosePrice_GTCAP[GTCAP_46_1:GTCAP_46_2])
C46 = np.save('AGIshortPrices.npy', ClosePrice_AGI[AGI_46_1:AGI_46_2])
E46 = np.save('ALIshortPrices.npy', ClosePrice_ALI[ALI_46_1:ALI_46_2])
ε46 = np.save('URCshortPrices.npy', ClosePrice_URC[URC_46_1:URC_46_2])
W46 = np.save('PGOLDshortPrices.npy', ClosePrice_PGOLD[PGOLD_46_1:PGOLD_46_2])

# import modules
import numpy as np
# import prices
L46 = np.load('GTCAPshortPrices.npy', allow_pickle=True).tolist()
C46 = np.load('AGIshortPrices.npy', allow_pickle=True).tolist()
E46 = np.load('ALIshortPrices.npy', allow_pickle=True).tolist()
ε46 = np.load('URCshortPrices.npy', allow_pickle=True).tolist()
W46 = np.load('PGOLDshortPrices.npy', allow_pickle=True).tolist()
```

```python
# extract spatio-temporal sequences
STC_46 = []
st_46 = []
n = 1
while n < len(L46):
    sc_46 = ''
    if L46[n] > L46[n-1]:
        sc_46 = sc_46 + 'L'
    if C46[n] > C46[n-1]:
        sc_46 = sc_46 + 'C'
    if E46[n] > E46[n-1]:
        sc_46 = sc_46 + 'E'
    if ε46[n] > ε46[n-1]:
        sc_46 = sc_46 + 'ε'
    if W46[n] > W46[n-1]:
        sc_46 = sc_46 + 'W'
    if len(sc_46) > 0:
        st_46.append(sc_46)
    else:
        STC_46.append(st_46)
        st_46 = []
    n = n + 1


# record the number of empty spatio-temporal sequences
nSTCempty_46 = STC_46.count([])
# keep only non-empty spatio-temporal sequences
STC_46 = [ st_46 for st_46 in STC_46 if len(st_46) > 0 ]

# record the length of the longest spatio-temporal sequence
lmin_46 = len(L46)
lmax_46 = 0
for st_46 in STC_46:
    if len(st_46) < lmin_46:
        lmin_46 = len(st_46)
    if len(st_46) > lmax_46:
        lmax_46 = len(st_46)
        
# restrict spatio-temporal sequences to length-5
STC5_46 = [ st_46 for st_46 in STC_46 if len(st_46) <= 5 ]

# unpack spatio-temporal sequences to temporal sequences
TC5_46 = []
for n in range(len(STC5_46)):
    tc_46 = [ u for u in STC5_46[n][0] ]
    for k in range(1, len(STC5_46[n])):
        tc_46 = [ u+v for u in tc_46 for v in STC5_46[n][k] ]
    TC5_46 = TC5_46 + tc_46
setTC5_46 = list(set(TC5_46))

# get empirical frequencies
fTC5_46 = [ TC5_46.count(setTC5_46[l]) for l in range(len(setTC5_46)) ]

# statistical testing against null model
np.random.seed(0)
fTC5null_46 = [ [] for l in range(len(setTC5_46)) ]
for s in range(100):
    STC5null_46 = [ list(st_46) for st_46 in STC5_46]
    for st_46 in STC5null_46:
        np.random.shuffle(st_46)
    TC5null_46= []
    for n in range(len(STC5null_46)):
        tc_46 = [ u for u in STC5null_46[n][0] ]
        for k in range(1, len(STC5null_46[n])):
            tc_46 = [ u+v for u in tc_46 for v in STC5null_46[n][k] ]
        TC5null_46= TC5null_46+ tc_46
    for l in range(len(setTC5_46)):
        if TC5null_46.count(setTC5_46[l]) > 0:
            fTC5null_46[l].append(TC5null_46.count(setTC5_46[l]))
        else:
            fTC5null_46[l].append(0)
```

```python
print("List of Statistically Significant Sequences at α = 0.05")

#statistical significance testing at p < 0.05 level
for l in range(len(setTC5_46)):
    if sum(freq >= fTC5_46[l] for freq in fTC5null_46[l]) < 5:
        print(setTC5_46[l], fTC5_46[l], sum(freq >= fTC5_46[l] for freq in fTC5null_46[l]))
```

```python
print("List of Statistically Significant Sequences at α = 0.01")

#statistical significance testing at p < 0.01 level
for l in range(len(setTC5_46)):
    if sum(freq >= fTC5_46[l] for freq in fTC5null_46[l]) < 1:
        print(setTC5_46[l], fTC5_46[l], sum(freq >= fTC5_46[l] for freq in fTC5null_46[l]))
```

```python
# extract spatio-temporal sequences
STC_49 = []
st_49 = []
n = 1
while n < len(L49):
    sc_49 = ''
    if L49[n] > L49[n-1]:
        sc_49 = sc_49 + 'L'
    if C49[n] > C49[n-1]:
        sc_49 = sc_49 + 'C'
    if Z49[n] > Z49[n-1]:
        sc_49 = sc_49 + 'Z'
    if T49[n] > T49[n-1]:
        sc_49 = sc_49 + 'T'
    if W49[n] > W49[n-1]:
        sc_49 = sc_49 + 'W'
    if len(sc_49) > 0:
        st_49.append(sc_49)
    else:
        STC_49.append(st_49)
        st_49 = []
    n = n + 1


# record the number of empty spatio-temporal sequences
nSTCempty_49 = STC_49.count([])
# keep only non-empty spatio-temporal sequences
STC_49 = [ st_49 for st_49 in STC_49 if len(st_49) > 0 ]

# record the length of the longest spatio-temporal sequence
lmin_49 = len(L49)
lmax_49 = 0
for st_49 in STC_49:
    if len(st_49) < lmin_49:
        lmin_49 = len(st_49)
    if len(st_49) > lmax_49:
        lmax_49 = len(st_49)
        
# restrict spatio-temporal sequences to length-5
STC5_49 = [ st_49 for st_49 in STC_49 if len(st_49) <= 5 ]

# unpack spatio-temporal sequences to temporal sequences
TC5_49 = []
for n in range(len(STC5_49)):
    tc_49 = [ u for u in STC5_49[n][0] ]
    for k in range(1, len(STC5_49[n])):
        tc_49 = [ u+v for u in tc_49 for v in STC5_49[n][k] ]
    TC5_49 = TC5_49 + tc_49
setTC5_49 = list(set(TC5_49))

# get empirical frequencies
fTC5_49 = [ TC5_49.count(setTC5_49[l]) for l in range(len(setTC5_49)) ]

# statistical testing against null model
np.random.seed(0)
fTC5null_49 = [ [] for l in range(len(setTC5_49)) ]
for s in range(100):
    STC5null_49 = [ list(st_49) for st_49 in STC5_49]
    for st_49 in STC5null_49:
        np.random.shuffle(st_49)
    TC5null_49= []
    for n in range(len(STC5null_49)):
        tc_49 = [ u for u in STC5null_49[n][0] ]
        for k in range(1, len(STC5null_49[n])):
            tc_49 = [ u+v for u in tc_49 for v in STC5null_49[n][k] ]
        TC5null_49= TC5null_49+ tc_49
    for l in range(len(setTC5_49)):
        if TC5null_49.count(setTC5_49[l]) > 0:
            fTC5null_49[l].append(TC5null_49.count(setTC5_49[l]))
        else:
            fTC5null_49[l].append(0)
```
