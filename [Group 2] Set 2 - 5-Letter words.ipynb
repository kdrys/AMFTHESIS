# 16 (L) GTCAP - (C) AGI - (β) SM - (T) MBT - (V) TEL

```python
#extract the overlapping dates

start_16 = max(min(Date_GTCAP), min(Date_AGI), min(Date_SM), min(Date_MBT), min(Date_TEL))
end_16 = min(max(Date_GTCAP), max(Date_AGI), max(Date_SM), max(Date_MBT), max(Date_TEL))
```

```python
#shorten the closing prices
GTCAP_16_1 = Date_GTCAP.index(start_16)
GTCAP_16_2 = Date_GTCAP.index(end_16)

AGI_16_1 = Date_AGI.index(start_16)
AGI_16_2 = Date_AGI.index(end_16)

SM_16_1 = Date_SM.index(start_16)
SM_16_2 = Date_SM.index(end_16)

MBT_16_1 = Date_MBT.index(start_16)
MBT_16_2 = Date_MBT.index(end_16)

TEL_16_1 = Date_TEL.index(start_16)
TEL_16_2 = Date_TEL.index(end_16)
```

```python
#save the shortened versions of closing prices
L16 = np.save('GTCAPshortPrices.npy', ClosePrice_GTCAP[GTCAP_16_1:GTCAP_16_2])
C16 = np.save('AGIshortPrices.npy', ClosePrice_AGI[AGI_16_1:AGI_16_2])
β16 = np.save('SMshortPrices.npy', ClosePrice_SM[SM_16_1:SM_16_2])
T16 = np.save('MBTshortPrices.npy', ClosePrice_MBT[MBT_16_1:MBT_16_2])
V16 = np.save('TELshortPrices.npy', ClosePrice_TEL[TEL_16_1:TEL_16_2])
```

```python
# import modules
import numpy as np
# import prices
L16 = np.load('GTCAPshortPrices.npy', allow_pickle=True).tolist()
C16 = np.load('AGIshortPrices.npy', allow_pickle=True).tolist()
β16 = np.load('SMshortPrices.npy', allow_pickle=True).tolist()
T16 = np.load('MBTshortPrices.npy', allow_pickle=True).tolist()
V16 = np.load('TELshortPrices.npy', allow_pickle=True).tolist()
```

```python
# extract spatio-temporal sequences
STC_16 = []
st_16 = []
n = 1
while n < len(L16):
    sc_16 = ''
    if L16[n] > L16[n-1]:
        sc_16 = sc_16 + 'L'
    if C16[n] > C16[n-1]:
        sc_16 = sc_16 + 'C'
    if β16[n] > β16[n-1]:
        sc_16 = sc_16 + 'β'
    if T16[n] > T16[n-1]:
        sc_16 = sc_16 + 'T'
    if V16[n] > V16[n-1]:
        sc_16 = sc_16 + 'V'
    if len(sc_16) > 0:
        st_16.append(sc_16)
    else:
        STC_16.append(st_16)
        st_16 = []
    n = n + 1
```

```python
# record the number of empty spatio-temporal sequences
nSTCempty_16 = STC_16.count([])
# keep only non-empty spatio-temporal sequences
STC_16 = [ st_16 for st_16 in STC_16 if len(st_16) > 0 ]

# record the length of the longest spatio-temporal sequence
lmin_16 = len(L16)
lmax_16 = 0
for st_16 in STC_16:
    if len(st_16) < lmin_16:
        lmin_16 = len(st_16)
    if len(st_16) > lmax_16:
        lmax_16 = len(st_16)
        
# restrict spatio-temporal sequences to length-5
STC5_16 = [ st_16 for st_16 in STC_16 if len(st_16) <= 5 ]

# unpack spatio-temporal sequences to temporal sequences
TC5_16 = []
for n in range(len(STC5_16)):
    tc_16 = [ u for u in STC5_16[n][0] ]
    for k in range(1, len(STC5_16[n])):
        tc_16 = [ u+v for u in tc_16 for v in STC5_16[n][k] ]
    TC5_16 = TC5_16 + tc_16
setTC5_16 = list(set(TC5_16))

# get empirical frequencies
fTC5_16 = [ TC5_16.count(setTC5_16[l]) for l in range(len(setTC5_16)) ]

# statistical testing against null model
np.random.seed(0)
fTC5null_16 = [ [] for l in range(len(setTC5_16)) ]
for s in range(100):
    STC5null_16 = [ list(st_16) for st_16 in STC5_16]
    for st_16 in STC5null_16:
        np.random.shuffle(st_16)
    TC5null_16= []
    for n in range(len(STC5null_16)):
        tc_16 = [ u for u in STC5null_16[n][0] ]
        for k in range(1, len(STC5null_16[n])):
            tc_16 = [ u+v for u in tc_16 for v in STC5null_16[n][k] ]
        TC5null_16= TC5null_16+ tc_16
    for l in range(len(setTC5_16)):
        if TC5null_16.count(setTC5_16[l]) > 0:
            fTC5null_16[l].append(TC5null_16.count(setTC5_16[l]))
        else:
            fTC5null_16[l].append(0)
```

```python
print("List of Statistically Significant Sequences at α = 0.05")

#statistical significance testing at p < 0.05 level
for l in range(len(setTC5_16)):
    if sum(freq >= fTC5_16[l] for freq in fTC5null_16[l]) < 5:
        print(setTC5_16[l], fTC5_16[l], sum(freq >= fTC5_16[l] for freq in fTC5null_16[l]))
```

```python
print("List of Statistically Significant Sequences at α = 0.01")

#statistical significance testing at p < 0.01 level
for l in range(len(setTC5_16)):
    if sum(freq >= fTC5_16[l] for freq in fTC5null_16[l]) < 1:
        print(setTC5_16[l], fTC5_16[l], sum(freq >= fTC5_16[l] for freq in fTC5null_16[l]))
```

```python
# record the number of empty spatio-temporal sequences
nSTCempty_27 = STC_27.count([])
# keep only non-empty spatio-temporal sequences
STC_27 = [ st_27 for st_27 in STC_27 if len(st_27) > 0 ]

# record the length of the longest spatio-temporal sequence
lmin_27 = len(L27)
lmax_27 = 0
for st_27 in STC_27:
    if len(st_27) < lmin_27:
        lmin_27 = len(st_27)
    if len(st_27) > lmax_27:
        lmax_27 = len(st_27)
        
# restrict spatio-temporal sequences to length-5
STC5_27 = [ st_27 for st_27 in STC_27 if len(st_27) <= 5 ]

# unpack spatio-temporal sequences to temporal sequences
TC5_27 = []
for n in range(len(STC5_27)):
    tc_27 = [ u for u in STC5_27[n][0] ]
    for k in range(1, len(STC5_27[n])):
        tc_27 = [ u+v for u in tc_27 for v in STC5_27[n][k] ]
    TC5_27 = TC5_27 + tc_27
setTC5_27 = list(set(TC5_27))

# get empirical frequencies
fTC5_27 = [ TC5_27.count(setTC5_27[l]) for l in range(len(setTC5_27)) ]

# statistical testing against null model
np.random.seed(0)
fTC5null_27 = [ [] for l in range(len(setTC5_27)) ]
for s in range(100):
    STC5null_27 = [ list(st_27) for st_27 in STC5_27]
    for st_27 in STC5null_27:
        np.random.shuffle(st_27)
    TC5null_27= []
    for n in range(len(STC5null_27)):
        tc_27 = [ u for u in STC5null_27[n][0] ]
        for k in range(1, len(STC5null_27[n])):
            tc_27 = [ u+v for u in tc_27 for v in STC5null_27[n][k] ]
        TC5null_27= TC5null_27+ tc_27
    for l in range(len(setTC5_27)):
        if TC5null_27.count(setTC5_27[l]) > 0:
            fTC5null_27[l].append(TC5null_27.count(setTC5_27[l]))
        else:
            fTC5null_27[l].append(0)
```

```python
print("List of Statistically Significant Sequences at α = 0.05")

#statistical significance testing at p < 0.05 level
for l in range(len(setTC5_27)):
    if sum(freq >= fTC5_27[l] for freq in fTC5null_27[l]) < 5:
        print(setTC5_27[l], fTC5_27[l], sum(freq >= fTC5_27[l] for freq in fTC5null_27[l]))
```
