# 91 (L) GTCAP - (β) SM - (V) TEL - (ε) URC - (W) PGOLD

```python
#extract the overlapping dates

start_91 = max(min(Date_GTCAP), min(Date_SM), min(Date_TEL), min(Date_URC), min(Date_PGOLD))
end_91 = min(max(Date_GTCAP), max(Date_SM), max(Date_TEL), max(Date_URC), max(Date_PGOLD))

#shorten the closing prices
GTCAP_91_1 = Date_GTCAP.index(start_91)
GTCAP_91_2 = Date_GTCAP.index(end_91)

SM_91_1 = Date_SM.index(start_91)
SM_91_2 = Date_SM.index(end_91)

TEL_91_1 = Date_TEL.index(start_91)
TEL_91_2 = Date_TEL.index(end_91)

URC_91_1 = Date_URC.index(start_91)
URC_91_2 = Date_URC.index(end_91)

PGOLD_91_1 = Date_PGOLD.index(start_91)
PGOLD_91_2 = Date_PGOLD.index(end_91)

#save the shortened versions of closing prices
L91 = np.save('GTCAPshortPrices.npy', ClosePrice_GTCAP[GTCAP_91_1:GTCAP_91_2])
β91 = np.save('SMshortPrices.npy', ClosePrice_SM[SM_91_1:SM_91_2])
V91 = np.save('TELshortPrices.npy', ClosePrice_TEL[TEL_91_1:TEL_91_2])
ε91 = np.save('URCshortPrices.npy', ClosePrice_URC[URC_91_1:URC_91_2])
W91 = np.save('PGOLDshortPrices.npy', ClosePrice_PGOLD[PGOLD_91_1:PGOLD_91_2])

# import modules
import numpy as np
# import prices
L91 = np.load('GTCAPshortPrices.npy', allow_pickle=True).tolist()
β91 = np.load('SMshortPrices.npy', allow_pickle=True).tolist()
V91 = np.load('TELshortPrices.npy', allow_pickle=True).tolist()
ε91 = np.load('URCshortPrices.npy', allow_pickle=True).tolist()
W91 = np.load('PGOLDshortPrices.npy', allow_pickle=True).tolist()
```

```python
# extract spatio-temporal sequences
STC_91 = []
st_91 = []
n = 1
while n < len(L91):
    sc_91 = ''
    if L91[n] > L91[n-1]:
        sc_91 = sc_91 + 'L'
    if β91[n] > β91[n-1]:
        sc_91 = sc_91 + 'β'
    if V91[n] > V91[n-1]:
        sc_91 = sc_91 + 'V'
    if ε91[n] > ε91[n-1]:
        sc_91 = sc_91 + 'ε'
    if W91[n] > W91[n-1]:
        sc_91 = sc_91 + 'W'
    if len(sc_91) > 0:
        st_91.append(sc_91)
    else:
        STC_91.append(st_91)
        st_91 = []
    n = n + 1


# record the number of empty spatio-temporal sequences
nSTCempty_91 = STC_91.count([])
# keep only non-empty spatio-temporal sequences
STC_91 = [ st_91 for st_91 in STC_91 if len(st_91) > 0 ]

# record the length of the longest spatio-temporal sequence
lmin_91 = len(L91)
lmax_91 = 0
for st_91 in STC_91:
    if len(st_91) < lmin_91:
        lmin_91 = len(st_91)
    if len(st_91) > lmax_91:
        lmax_91 = len(st_91)
        
# restrict spatio-temporal sequences to length-5
STC5_91 = [ st_91 for st_91 in STC_91 if len(st_91) <= 5 ]

# unpack spatio-temporal sequences to temporal sequences
TC5_91 = []
for n in range(len(STC5_91)):
    tc_91 = [ u for u in STC5_91[n][0] ]
    for k in range(1, len(STC5_91[n])):
        tc_91 = [ u+v for u in tc_91 for v in STC5_91[n][k] ]
    TC5_91 = TC5_91 + tc_91
setTC5_91 = list(set(TC5_91))

# get empirical frequencies
fTC5_91 = [ TC5_91.count(setTC5_91[l]) for l in range(len(setTC5_91)) ]

# statistical testing against null model
np.random.seed(0)
fTC5null_91 = [ [] for l in range(len(setTC5_91)) ]
for s in range(100):
    STC5null_91 = [ list(st_91) for st_91 in STC5_91]
    for st_91 in STC5null_91:
        np.random.shuffle(st_91)
    TC5null_91= []
    for n in range(len(STC5null_91)):
        tc_91 = [ u for u in STC5null_91[n][0] ]
        for k in range(1, len(STC5null_91[n])):
            tc_91 = [ u+v for u in tc_91 for v in STC5null_91[n][k] ]
        TC5null_91= TC5null_91+ tc_91
    for l in range(len(setTC5_91)):
        if TC5null_91.count(setTC5_91[l]) > 0:
            fTC5null_91[l].append(TC5null_91.count(setTC5_91[l]))
        else:
            fTC5null_91[l].append(0)
```

```python
print("List of Statistically Significant Sequences at α = 0.05")

#statistical significance testing at p < 0.05 level
for l in range(len(setTC5_91)):
    if sum(freq >= fTC5_91[l] for freq in fTC5null_91[l]) < 5:
        print(setTC5_91[l])
```

```python

print("List of Statistically Significant Sequences at α = 0.01")

#statistical significance testing at p < 0.01 level
for l in range(len(setTC5_91)):
    if sum(freq >= fTC5_91[l] for freq in fTC5null_91[l]) < 1:
        print(setTC5_91[l])
```

```python
# extract spatio-temporal sequences
STC_102 = []
st_102 = []
n = 1
while n < len(L102):
    sc_102 = ''
    if L102[n] > L102[n-1]:
        sc_102 = sc_102 + 'L'
    if Q102[n] > Q102[n-1]:
        sc_102 = sc_102 + 'Q'
    if Z102[n] > Z102[n-1]:
        sc_102 = sc_102 + 'Z'
    if T102[n] > T102[n-1]:
        sc_102 = sc_102 + 'T'
    if V102[n] > V102[n-1]:
        sc_102 = sc_102 + 'V'
    if len(sc_102) > 0:
        st_102.append(sc_102)
    else:
        STC_102.append(st_102)
        st_102 = []
    n = n + 1


# record the number of empty spatio-temporal sequences
nSTCempty_102 = STC_102.count([])
# keep only non-empty spatio-temporal sequences
STC_102 = [ st_102 for st_102 in STC_102 if len(st_102) > 0 ]

# record the length of the longest spatio-temporal sequence
lmin_102 = len(L102)
lmax_102 = 0
for st_102 in STC_102:
    if len(st_102) < lmin_102:
        lmin_102 = len(st_102)
    if len(st_102) > lmax_102:
        lmax_102 = len(st_102)
        
# restrict spatio-temporal sequences to length-5
STC5_102 = [ st_102 for st_102 in STC_102 if len(st_102) <= 5 ]

# unpack spatio-temporal sequences to temporal sequences
TC5_102 = []
for n in range(len(STC5_102)):
    tc_102 = [ u for u in STC5_102[n][0] ]
    for k in range(1, len(STC5_102[n])):
        tc_102 = [ u+v for u in tc_102 for v in STC5_102[n][k] ]
    TC5_102 = TC5_102 + tc_102
setTC5_102 = list(set(TC5_102))

# get empirical frequencies
fTC5_102 = [ TC5_102.count(setTC5_102[l]) for l in range(len(setTC5_102)) ]

# statistical testing against null model
np.random.seed(0)
fTC5null_102 = [ [] for l in range(len(setTC5_102)) ]
for s in range(102):
    STC5null_102 = [ list(st_102) for st_102 in STC5_102]
    for st_102 in STC5null_102:
        np.random.shuffle(st_102)
    TC5null_102= []
    for n in range(len(STC5null_102)):
        tc_102 = [ u for u in STC5null_102[n][0] ]
        for k in range(1, len(STC5null_102[n])):
            tc_102 = [ u+v for u in tc_102 for v in STC5null_102[n][k] ]
        TC5null_102= TC5null_102+ tc_102
    for l in range(len(setTC5_102)):
        if TC5null_102.count(setTC5_102[l]) > 0:
            fTC5null_102[l].append(TC5null_102.count(setTC5_102[l]))
        else:
            fTC5null_102[l].append(0)
```

```python
print("List of Statistically Significant Sequences at α = 0.05")

#statistical significance testing at p < 0.05 level
for l in range(len(setTC5_102)):
    if sum(freq >= fTC5_102[l] for freq in fTC5null_102[l]) < 5:
        print(setTC5_102[l])
```

```python
print("List of Statistically Significant Sequences at α = 0.01")

#statistical significance testing at p < 0.01 level
for l in range(len(setTC5_102)):
    if sum(freq >= fTC5_102[l] for freq in fTC5null_102[l]) < 1:
        print(setTC5_102[l])
```
