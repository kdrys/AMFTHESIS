```python
# import modules
import numpy as np
# import prices

B76 = np.load('APshortPrices.npy', allow_pickle=True).tolist() 
F76 = np.load('BPIshortPrices.npy', allow_pickle=True).tolist() 
J76 = np.load('FGENshortPrices.npy', allow_pickle=True).tolist() 
N76 = np.load('JGSshortPrices.npy', allow_pickle=True).tolist() 
R76 = np.load('MEGshortPrices.npy', allow_pickle=True).tolist()
```

```python
# extract spatio-temporal sequences
STC_76 = []
st_76 = []
n = 1
while n < len(B76):
    sc_76 = ''
    if B76[n] > B76[n-1]:
        sc_76 = sc_76 + 'B'
    if F76[n] > F76[n-1]:
        sc_76 = sc_76 + 'F'
    if J76[n] > J76[n-1]:
        sc_76 = sc_76 + 'J'
    if N76[n] > N76[n-1]:
        sc_76 = sc_76 + 'N'
    if R76[n] > R76[n-1]:
        sc_76 = sc_76 + 'R'
    if len(sc_76) > 0:
        st_76.append(sc_76)
    else:
        STC_76.append(st_76)
        st_76 = []
    n = n + 1
```

```python
# record the number of empty spatio-temporal sequences
nSTCempty_76 = STC_76.count([])
# keep only non-empty spatio-temporal sequences
STC_76 = [ st_76 for st_76 in STC_76 if len(st_76) > 0 ]

# record the length of the longest spatio-temporal sequence
lmin_76 = len(B76)
lmax_76 = 0
for st_76 in STC_76:
    if len(st_76) < lmin_76:
        lmin_76 = len(st_76)
    if len(st_76) > lmax_76:
        lmax_76 = len(st_76)
        
# restrict spatio-temporal sequences to length-5
STC5_76 = [ st_76 for st_76 in STC_76 if len(st_76) <= 5 ]

# unpackk spatio-temporal sequences to temporal sequences
TC5_76 = []
for n in range(len(STC5_76)):
    tc_76 = [ u for u in STC5_76[n][0] ]
    for k in range(1, len(STC5_76[n])):
        tc_76 = [ u+v for u in tc_76 for v in STC5_76[n][k] ]
    TC5_76 = TC5_76 + tc_76
setTC5_76 = list(set(TC5_76))

# get empirical frequencies
fTC5_76 = [ TC5_76.count(setTC5_76[l]) for l in range(len(setTC5_76)) ]

# statistical testing against null model
np.random.seed(0)
fTC5null_76 = [ [] for l in range(len(setTC5_76)) ]
for s in range(100):
    STC5null_76 = [ list(st_76) for st_76 in STC5_76]
    for st_76 in STC5null_76:
        np.random.shuffle(st_76)
    TC5null_76= []
    for n in range(len(STC5null_76)):
        tc_76 = [ u for u in STC5null_76[n][0] ]
        for k in range(1, len(STC5null_76[n])):
            tc_76 = [ u+v for u in tc_76 for v in STC5null_76[n][k] ]
        TC5null_76= TC5null_76+ tc_76
    for l in range(len(setTC5_76)):
        if TC5null_76.count(setTC5_76[l]) > 0:
            fTC5null_76[l].append(TC5null_76.count(setTC5_76[l]))
        else:
            fTC5null_76[l].append(0)
```

```python
print("List of Statistically Significant Sequences at R = 0.05")

#statistical significance testing at p < 0.05 level
for l in range(len(setTC5_76)):
     if sum(freq >= fTC5_76[l] for freq in fTC5null_76[l]) < 5:
#         print(setTC5_76[l], fTC5_76[l], sum(freq >= fTC5_76[l] for freq in fTC5null_76[l]))
        print(setTC5_76[l])
```

```python
print("List of Statistically Significant Sequences at R = 0.01")

#statistical significance testing at p < 0.01 level
for l in range(len(setTC5_76)):
     if sum(freq >= fTC5_76[l] for freq in fTC5null_76[l]) < 1:
#         print(setTC5_76[l], fTC5_76[l], sum(freq >= fTC5_76[l] for freq in fTC5null_76[l]))
        print(setTC5_76[l])
```
