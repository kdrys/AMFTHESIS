##  1 AEV

```python
# Aboitiz Equity Ventures Inc.
# open the csv file
AEV = open('AEV.csv')

# initialize the lists
Date_AEV = []
ClosePrice_AEV = []

# skip the header line
AEV.readline()
#The readline() method returns one line from the file. 

# go through the file line by line
for line in AEV:
    sline = line.strip().split(',') #list type
    #The strip() method removes the whitespaces or specific characters 
    #from the string at the beginning and end of the string.
    
    #The split() method splits a string into a list. 
    
    #Use the indexes in the sline list to append the values corresponding to the defined lists
    Date_AEV.append(sline[0])
    ClosePrice_AEV.append(float(sline[1]))

    #returns an error when it encountered missing values
    
# close file
AEV.close()

# import numpy and datetime
import numpy as np
import datetime

# convert dates into datetime structure
for n in range(len(Date_AEV)):
    #The strptime() method creates a datetime object from the string at index 'n' in the *Date* list.
    #The .date() function extracts only the date part and assigns it back to the *Date* list. 
    Date_AEV[n] = datetime.datetime.strptime(Date_AEV[n], '%m/%d/%y').date()
    
# save Date and ClosePrice
np.save('AEVDate.npy', Date_AEV)
np.save('AEVClosePrice.npy', ClosePrice_AEV)

#import modules
import numpy as np
import matplotlib.pyplot as plt

# import AEV Closing Prices
A = np.load('AEVClosePrice.npy').tolist()
A

# initialize array of one-letter strings
SA = []

# initialize empty string and counter
sA = '' #empty string
nA = 1 #counter
# while loop
while nA < len(A):
    if A[nA] > A[nA-1]: #checks if the price increased?
        sA = sA + 'A'
    else:
        SA.append(sA) #appending A to the strings of A
        sA = ''
    nA = nA + 1

# frequencies of words up to length-5
FA = [SA.count('A'), SA.count('AA'), SA.count('AAA'), SA.count('AAAA'), SA.count('AAAAA')]
#Count() is a Python built-in function that returns the number of times an object appears in a list. 
FA

# plot frequency as function of length
fig = plt.figure()
fig.subplots_adjust(bottom=0.15)
fig.subplots_adjust(top=0.925)
fig.subplots_adjust(left=0.15)
fig.subplots_adjust(right=0.95)
plt.semilogy([1,2,3,4,5], FA, 'o-', linewidth=2, markersize=10)
plt.xticks([1,2,3,4,5], [r'$1$', r'$2$', r'$3$', r'$4$', r'$5$'], fontsize=18)
plt.xlabel('word length', fontsize=18)
plt.yticks([5e1,1e2,1e3,2e3], ['', r'$10ˆ2$', r'$10ˆ3$', ''], fontsize=18)
plt.ylabel('frequency', fontsize=18)
plt.title('(1) AEV', fontsize=18)
plt.savefig('freq_AEV.pdf')
plt.show()
```
