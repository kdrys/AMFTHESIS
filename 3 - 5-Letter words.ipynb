```python
#extract the overlapping dates

start_31 = max(min(Date_GTCAP), min(Date_AGI), min(Date_MER), min(Date_MBT), min(Date_TEL))
end_31 = min(max(Date_GTCAP), max(Date_AGI), max(Date_MER), max(Date_MBT), max(Date_TEL))
```

```python
#shorten the closing prices
GTCAP_31_1 = Date_GTCAP.index(start_31)
GTCAP_31_2 = Date_GTCAP.index(end_31)

AGI_31_1 = Date_AGI.index(start_31)
AGI_31_2 = Date_AGI.index(end_31)

MER_31_1 = Date_MER.index(start_31)
MER_31_2 = Date_MER.index(end_31)

MBT_31_1 = Date_MBT.index(start_31)
MBT_31_2 = Date_MBT.index(end_31)

TEL_31_1 = Date_TEL.index(start_31)
TEL_31_2 = Date_TEL.index(end_31)
```

```python
#save the shortened versions of closing prices
L31 = np.save('GTCAPshortPrices.npy', ClosePrice_GTCAP[GTCAP_31_1:GTCAP_31_2])
C31 = np.save('AGIshortPrices.npy', ClosePrice_AGI[AGI_31_1:AGI_31_2])
Q31 = np.save('MERshortPrices.npy', ClosePrice_MER[MER_31_1:MER_31_2])
T31 = np.save('MBTshortPrices.npy', ClosePrice_MBT[MBT_31_1:MBT_31_2])
V31 = np.save('TELshortPrices.npy', ClosePrice_TEL[TEL_31_1:TEL_31_2])
```

```python
# import prices
L31 = np.load('GTCAPshortPrices.npy', allow_pickle=True).tolist()
C31 = np.load('AGIshortPrices.npy', allow_pickle=True).tolist()
Q31 = np.load('MERshortPrices.npy', allow_pickle=True).tolist()
T31 = np.load('MBTshortPrices.npy', allow_pickle=True).tolist()
V31 = np.load('TELshortPrices.npy', allow_pickle=True).tolist()
```

```python
# extract spatio-temporal sequences
STC_31 = []
st_31 = []
n = 1
while n < len(L31):
    sc_31 = ''
    if L31[n] > L31[n-1]:
        sc_31 = sc_31 + 'L'
    if C31[n] > C31[n-1]:
        sc_31 = sc_31 + 'C'
    if Q31[n] > Q31[n-1]:
        sc_31 = sc_31 + 'Q'
    if T31[n] > T31[n-1]:
        sc_31 = sc_31 + 'T'
    if V31[n] > V31[n-1]:
        sc_31 = sc_31 + 'V'
    if len(sc_31) > 0:
        st_31.append(sc_31)
    else:
        STC_31.append(st_31)
        st_31 = []
    n = n + 1
```

```python
# record the number of empty spatio-temporal sequences
nSTCempty_31 = STC_31.count([])
# keep only non-empty spatio-temporal sequences
STC_31 = [ st_31 for st_31 in STC_31 if len(st_31) > 0 ]

# record the length of the longest spatio-temporal sequence
lmin_31 = len(L31)
lmax_31 = 0
for st_31 in STC_31:
    if len(st_31) < lmin_31:
        lmin_31 = len(st_31)
    if len(st_31) > lmax_31:
        lmax_31 = len(st_31)
        
# restrict spatio-temporal sequences to length-5
STC5_31 = [ st_31 for st_31 in STC_31 if len(st_31) <= 5 ]

# unpack spatio-temporal sequences to temporal sequences
TC5_31 = []
for n in range(len(STC5_31)):
    tc_31 = [ u for u in STC5_31[n][0] ]
    for k in range(1, len(STC5_31[n])):
        tc_31 = [ u+v for u in tc_31 for v in STC5_31[n][k] ]
    TC5_31 = TC5_31 + tc_31
setTC5_31 = list(set(TC5_31))

# get empirical frequencies
fTC5_31 = [ TC5_31.count(setTC5_31[l]) for l in range(len(setTC5_31)) ]

# statistical testing against null model
np.random.seed(0)
fTC5null_31 = [ [] for l in range(len(setTC5_31)) ]
for s in range(100):
    STC5null_31 = [ list(st_31) for st_31 in STC5_31]
    for st_31 in STC5null_31:
        np.random.shuffle(st_31)
    TC5null_31= []
    for n in range(len(STC5null_31)):
        tc_31 = [ u for u in STC5null_31[n][0] ]
        for k in range(1, len(STC5null_31[n])):
            tc_31 = [ u+v for u in tc_31 for v in STC5null_31[n][k] ]
        TC5null_31= TC5null_31+ tc_31
    for l in range(len(setTC5_31)):
        if TC5null_31.count(setTC5_31[l]) > 0:
            fTC5null_31[l].append(TC5null_31.count(setTC5_31[l]))
        else:
            fTC5null_31[l].append(0)
```

```python
print("List of Statistically Significant Sequences at α = 0.05")

#statistical significance testing at p < 0.01 level
for l in range(len(setTC5_31)):
    if sum(freq >= fTC5_31[l] for freq in fTC5null_31[l]) < 5:
        print(setTC5_31[l], fTC5_31[l], sum(freq >= fTC5_31[l] for freq in fTC5null_31[l]))
```

```python
print("List of Statistically Significant Sequences at α = 0.01")

#statistical significance testing at p < 0.01 level
for l in range(len(setTC5_31)):
    if sum(freq >= fTC5_31[l] for freq in fTC5null_31[l]) < 1:
        print(setTC5_31[l], fTC5_31[l], sum(freq >= fTC5_31[l] for freq in fTC5null_31[l]))
```
