```python
#shorten the closing prices
AP_61_1 = Date_AP.index(start_61)
AP_61_2 = Date_AP.index(end_61)

BPI_61_1 = Date_BPI.index(start_61)
BPI_61_2 = Date_BPI.index(end_61)

BLOOM_61_1 = Date_BLOOM.index(start_61)
BLOOM_61_2 = Date_BLOOM.index(end_61)

FGEN_61_1 = Date_FGEN.index(start_61)
FGEN_61_2 = Date_FGEN.index(end_61)

SCC_61_1 = Date_SCC.index(start_61)
SCC_61_2 = Date_SCC.index(end_61)
```

```python
#save the shortened versions of closing prices
B61 = np.save('APshortPrices.npy', ClosePrice_AP[AP_61_1:AP_61_2])
F61 = np.save('BPIshortPrices.npy', ClosePrice_BPI[BPI_61_1:BPI_61_2])
H61 = np.save('BLOOMshortPrices.npy', ClosePrice_BLOOM[BLOOM_61_1:BLOOM_61_2])
J61 = np.save('FGENshortPrices.npy', ClosePrice_FGEN[FGEN_61_1:FGEN_61_2])
α61 = np.save('SCCshortPrices.npy', ClosePrice_SCC[SCC_61_1:SCC_61_2])
```

```python
# import modules
import numpy as np
# import prices

B61 = np.load('APshortPrices.npy', allow_pickle=True).tolist() 
F61 = np.load('BPIshortPrices.npy', allow_pickle=True).tolist() 
H61 = np.load('BLOOMshortPrices.npy', allow_pickle=True).tolist() 
J61 = np.load('FGENshortPrices.npy', allow_pickle=True).tolist() 
α61 = np.load('SCCshortPrices.npy', allow_pickle=True).tolist()
```

```python
# extract spatio-temporal sequences
STC_61 = []
st_61 = []
n = 1
while n < len(B61):
    sc_61 = ''
    if B61[n] > B61[n-1]:
        sc_61 = sc_61 + 'B'
    if F61[n] > F61[n-1]:
        sc_61 = sc_61 + 'F'
    if H61[n] > H61[n-1]:
        sc_61 = sc_61 + 'H'
    if J61[n] > J61[n-1]:
        sc_61 = sc_61 + 'J'
    if α61[n] > α61[n-1]:
        sc_61 = sc_61 + 'α'
    if len(sc_61) > 0:
        st_61.append(sc_61)
    else:
        STC_61.append(st_61)
        st_61 = []
    n = n + 1
```

```python
# record the number of empty spatio-temporal sequences
nSTCempty_61 = STC_61.count([])
# keep only non-empty spatio-temporal sequences
STC_61 = [ st_61 for st_61 in STC_61 if len(st_61) > 0 ]

# record the length of the longest spatio-temporal sequence
lmin_61 = len(B61)
lmax_61 = 0
for st_61 in STC_61:
    if len(st_61) < lmin_61:
        lmin_61 = len(st_61)
    if len(st_61) > lmax_61:
        lmax_61 = len(st_61)
        
# restrict spatio-temporal sequences to length-5
STC5_61 = [ st_61 for st_61 in STC_61 if len(st_61) <= 5 ]

# unpBPIk spatio-temporal sequences to temporal sequences
TC5_61 = []
for n in range(len(STC5_61)):
    tc_61 = [ u for u in STC5_61[n][0] ]
    for k in range(1, len(STC5_61[n])):
        tc_61 = [ u+v for u in tc_61 for v in STC5_61[n][k] ]
    TC5_61 = TC5_61 + tc_61
setTC5_61 = list(set(TC5_61))

# get empirical frequencies
fTC5_61 = [ TC5_61.count(setTC5_61[l]) for l in range(len(setTC5_61)) ]

# statistical testing against null model
np.random.seed(0)
fTC5null_61 = [ [] for l in range(len(setTC5_61)) ]
for s in range(100):
    STC5null_61 = [ list(st_61) for st_61 in STC5_61]
    for st_61 in STC5null_61:
        np.random.shuffle(st_61)
    TC5null_61= []
    for n in range(len(STC5null_61)):
        tc_61 = [ u for u in STC5null_61[n][0] ]
        for k in range(1, len(STC5null_61[n])):
            tc_61 = [ u+v for u in tc_61 for v in STC5null_61[n][k] ]
        TC5null_61= TC5null_61+ tc_61
    for l in range(len(setTC5_61)):
        if TC5null_61.count(setTC5_61[l]) > 0:
            fTC5null_61[l].append(TC5null_61.count(setTC5_61[l]))
        else:
            fTC5null_61[l].append(0)
```

```python
print("List of Statistically Significant Sequences at R = 0.05")

#statistical significance testing at p < 0.05 level
for l in range(len(setTC5_61)):
     if sum(freq >= fTC5_61[l] for freq in fTC5null_61[l]) < 5:
#         print(setTC5_61[l], fTC5_61[l], sum(freq >= fTC5_61[l] for freq in fTC5null_61[l]))
        print(setTC5_61[l])
```

```python
print("List of Statistically Significant Sequences at R = 0.01")

#statistical significance testing at p < 0.01 level
for l in range(len(setTC5_61)):
     if sum(freq >= fTC5_61[l] for freq in fTC5null_61[l]) < 1:
#         print(setTC5_61[l], fTC5_61[l], sum(freq >= fTC5_61[l] for freq in fTC5null_61[l]))
        print(setTC5_61[l])
```
