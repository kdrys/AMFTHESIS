# 151 (C) AGI - (β) SM - (E) ALI - (ε) URC - (W) PGOLD

```python
#extract the overlapping dates

start_151 = max(min(Date_AGI), min(Date_SM), min(Date_ALI), min(Date_URC), min(Date_PGOLD))
end_151 = min(max(Date_AGI), max(Date_SM), max(Date_ALI), max(Date_URC), max(Date_PGOLD))

#shorten the closing prices
AGI_151_1 = Date_AGI.index(start_151)
AGI_151_2 = Date_AGI.index(end_151)

SM_151_1 = Date_SM.index(start_151)
SM_151_2 = Date_SM.index(end_151)

ALI_151_1 = Date_ALI.index(start_151)
ALI_151_2 = Date_ALI.index(end_151)

URC_151_1 = Date_URC.index(start_151)
URC_151_2 = Date_URC.index(end_151)

PGOLD_151_1 = Date_PGOLD.index(start_151)
PGOLD_151_2 = Date_PGOLD.index(end_151)

#save the shortened versions of closing prices
C151 = np.save('AGIshortPrices.npy', ClosePrice_AGI[AGI_151_1:AGI_151_2])
β151 = np.save('SMshortPrices.npy', ClosePrice_SM[SM_151_1:SM_151_2])
E151 = np.save('ALIshortPrices.npy', ClosePrice_ALI[ALI_151_1:ALI_151_2])
ε151 = np.save('URCshortPrices.npy', ClosePrice_URC[URC_151_1:URC_151_2])
W151 = np.save('PGOLDshortPrices.npy', ClosePrice_PGOLD[PGOLD_151_1:PGOLD_151_2])

# import modules
import numpy as np
# import prices
C151 = np.load('AGIshortPrices.npy', allow_pickle=True).tolist()
β151 = np.load('SMshortPrices.npy', allow_pickle=True).tolist()
E151 = np.load('ALIshortPrices.npy', allow_pickle=True).tolist()
ε151 = np.load('URCshortPrices.npy', allow_pickle=True).tolist()
W151 = np.load('PGOLDshortPrices.npy', allow_pickle=True).tolist()
```

```python
# extract spatio-temporal sequences
STC_151 = []
st_151 = []
n = 1
while n < len(C151):
    sc_151 = ''
    if C151[n] > C151[n-1]:
        sc_151 = sc_151 + 'C'
    if β151[n] > β151[n-1]:
        sc_151 = sc_151 + 'β'
    if E151[n] > E151[n-1]:
        sc_151 = sc_151 + 'E'
    if ε151[n] > ε151[n-1]:
        sc_151 = sc_151 + 'ε'
    if W151[n] > W151[n-1]:
        sc_151 = sc_151 + 'W'
    if len(sc_151) > 0:
        st_151.append(sc_151)
    else:
        STC_151.append(st_151)
        st_151 = []
    n = n + 1


# record the number of empty spatio-temporal sequences
nSTCempty_151 = STC_151.count([])
# keep only non-empty spatio-temporal sequences
STC_151 = [ st_151 for st_151 in STC_151 if len(st_151) > 0 ]

# record the length of the longest spatio-temporal sequence
lmin_151 = len(C151)
lmax_151 = 0
for st_151 in STC_151:
    if len(st_151) < lmin_151:
        lmin_151 = len(st_151)
    if len(st_151) > lmax_151:
        lmax_151 = len(st_151)
        
# restrict spatio-temporal sequences to length-5
STC5_151 = [ st_151 for st_151 in STC_151 if len(st_151) <= 5 ]

# unpack spatio-temporal sequences to temporal sequences
TC5_151 = []
for n in range(len(STC5_151)):
    tc_151 = [ u for u in STC5_151[n][0] ]
    for k in range(1, len(STC5_151[n])):
        tc_151 = [ u+v for u in tc_151 for v in STC5_151[n][k] ]
    TC5_151 = TC5_151 + tc_151
setTC5_151 = list(set(TC5_151))

# get empirical frequencies
fTC5_151 = [ TC5_151.count(setTC5_151[l]) for l in range(len(setTC5_151)) ]

# statistical testing against null model
np.random.seed(0)
fTC5null_151 = [ [] for l in range(len(setTC5_151)) ]
for s in range(151):
    STC5null_151 = [ list(st_151) for st_151 in STC5_151]
    for st_151 in STC5null_151:
        np.random.shuffle(st_151)
    TC5null_151= []
    for n in range(len(STC5null_151)):
        tc_151 = [ u for u in STC5null_151[n][0] ]
        for k in range(1, len(STC5null_151[n])):
            tc_151 = [ u+v for u in tc_151 for v in STC5null_151[n][k] ]
        TC5null_151= TC5null_151+ tc_151
    for l in range(len(setTC5_151)):
        if TC5null_151.count(setTC5_151[l]) > 0:
            fTC5null_151[l].append(TC5null_151.count(setTC5_151[l]))
        else:
            fTC5null_151[l].append(0)
```

```python
print("List of Statistically Significant Sequences at α = 0.05")

#statistical significance testing at p < 0.05 level
for l in range(len(setTC5_151)):
    if sum(freq >= fTC5_151[l] for freq in fTC5null_151[l]) < 5:
        print(setTC5_151[l], fTC5_151[l], sum(freq >= fTC5_151[l] for freq in fTC5null_151[l]))
```

```python
print("List of Statistically Significant Sequences at α = 0.01")

#statistical significance testing at p < 0.01 level
for l in range(len(setTC5_151)):
    if sum(freq >= fTC5_151[l] for freq in fTC5null_151[l]) < 1:
        print(setTC5_151[l], fTC5_151[l], sum(freq >= fTC5_151[l] for freq in fTC5null_151[l]))
```

```python
# extract spatio-temporal sequences
STC_154 = []
st_154 = []
n = 1
while n < len(C154):
    sc_154 = ''
    if C154[n] > C154[n-1]:
        sc_154 = sc_154 + 'C'
    if β154[n] > β154[n-1]:
        sc_154 = sc_154 + 'β'
    if Z154[n] > Z154[n-1]:
        sc_154 = sc_154 + 'Z'
    if T154[n] > T154[n-1]:
        sc_154 = sc_154 + 'T'
    if W154[n] > W154[n-1]:
        sc_154 = sc_154 + 'W'
    if len(sc_154) > 0:
        st_154.append(sc_154)
    else:
        STC_154.append(st_154)
        st_154 = []
    n = n + 1


# record the number of empty spatio-temporal sequences
nSTCempty_154 = STC_154.count([])
# keep only non-empty spatio-temporal sequences
STC_154 = [ st_154 for st_154 in STC_154 if len(st_154) > 0 ]

# record the length of the longest spatio-temporal sequence
lmin_154 = len(C154)
lmax_154 = 0
for st_154 in STC_154:
    if len(st_154) < lmin_154:
        lmin_154 = len(st_154)
    if len(st_154) > lmax_154:
        lmax_154 = len(st_154)
        
# restrict spatio-temporal sequences to length-5
STC5_154 = [ st_154 for st_154 in STC_154 if len(st_154) <= 5 ]

# unpack spatio-temporal sequences to temporal sequences
TC5_154 = []
for n in range(len(STC5_154)):
    tc_154 = [ u for u in STC5_154[n][0] ]
    for k in range(1, len(STC5_154[n])):
        tc_154 = [ u+v for u in tc_154 for v in STC5_154[n][k] ]
    TC5_154 = TC5_154 + tc_154
setTC5_154 = list(set(TC5_154))

# get empirical frequencies
fTC5_154 = [ TC5_154.count(setTC5_154[l]) for l in range(len(setTC5_154)) ]

# statistical testing against null model
np.random.seed(0)
fTC5null_154 = [ [] for l in range(len(setTC5_154)) ]
for s in range(154):
    STC5null_154 = [ list(st_154) for st_154 in STC5_154]
    for st_154 in STC5null_154:
        np.random.shuffle(st_154)
    TC5null_154= []
    for n in range(len(STC5null_154)):
        tc_154 = [ u for u in STC5null_154[n][0] ]
        for k in range(1, len(STC5null_154[n])):
            tc_154 = [ u+v for u in tc_154 for v in STC5null_154[n][k] ]
        TC5null_154= TC5null_154+ tc_154
    for l in range(len(setTC5_154)):
        if TC5null_154.count(setTC5_154[l]) > 0:
            fTC5null_154[l].append(TC5null_154.count(setTC5_154[l]))
        else:
            fTC5null_154[l].append(0)
```

```python
print("List of Statistically Significant Sequences at α = 0.05")

#statistical significance testing at p < 0.05 level
for l in range(len(setTC5_154)):
    if sum(freq >= fTC5_154[l] for freq in fTC5null_154[l]) < 5:
        print(setTC5_154[l], fTC5_154[l], sum(freq >= fTC5_154[l] for freq in fTC5null_154[l]))
```

```python
# extract spatio-temporal sequences
STC_158 = []
st_158 = []
n = 1
while n < len(C158):
    sc_158 = ''
    if C158[n] > C158[n-1]:
        sc_158 = sc_158 + 'C'
    if β158[n] > β158[n-1]:
        sc_158 = sc_158 + 'β'
    if T158[n] > T158[n-1]:
        sc_158 = sc_158 + 'T'
    if V158[n] > V158[n-1]:
        sc_158 = sc_158 + 'V'
    if ε158[n] > ε158[n-1]:
        sc_158 = sc_158 + 'ε'
    if len(sc_158) > 0:
        st_158.append(sc_158)
    else:
        STC_158.append(st_158)
        st_158 = []
    n = n + 1


# record the number of empty spatio-temporal sequences
nSTCempty_158 = STC_158.count([])
# keep only non-empty spatio-temporal sequences
STC_158 = [ st_158 for st_158 in STC_158 if len(st_158) > 0 ]

# record the length of the longest spatio-temporal sequence
lmin_158 = len(C158)
lmax_158 = 0
for st_158 in STC_158:
    if len(st_158) < lmin_158:
        lmin_158 = len(st_158)
    if len(st_158) > lmax_158:
        lmax_158 = len(st_158)
        
# restrict spatio-temporal sequences to length-5
STC5_158 = [ st_158 for st_158 in STC_158 if len(st_158) <= 5 ]

# unpack spatio-temporal sequences to temporal sequences
TC5_158 = []
for n in range(len(STC5_158)):
    tc_158 = [ u for u in STC5_158[n][0] ]
    for k in range(1, len(STC5_158[n])):
        tc_158 = [ u+v for u in tc_158 for v in STC5_158[n][k] ]
    TC5_158 = TC5_158 + tc_158
setTC5_158 = list(set(TC5_158))

# get empirical frequencies
fTC5_158 = [ TC5_158.count(setTC5_158[l]) for l in range(len(setTC5_158)) ]

# statistical testing against null model
np.random.seed(0)
fTC5null_158 = [ [] for l in range(len(setTC5_158)) ]
for s in range(158):
    STC5null_158 = [ list(st_158) for st_158 in STC5_158]
    for st_158 in STC5null_158:
        np.random.shuffle(st_158)
    TC5null_158= []
    for n in range(len(STC5null_158)):
        tc_158 = [ u for u in STC5null_158[n][0] ]
        for k in range(1, len(STC5null_158[n])):
            tc_158 = [ u+v for u in tc_158 for v in STC5null_158[n][k] ]
        TC5null_158= TC5null_158+ tc_158
    for l in range(len(setTC5_158)):
        if TC5null_158.count(setTC5_158[l]) > 0:
            fTC5null_158[l].append(TC5null_158.count(setTC5_158[l]))
        else:
            fTC5null_158[l].append(0)
```

```python
# extract spatio-temporal sequences
STC_162 = []
st_162 = []
n = 1
while n < len(C162):
    sc_162 = ''
    if C162[n] > C162[n-1]:
        sc_162 = sc_162 + 'C'
    if Q162[n] > Q162[n-1]:
        sc_162 = sc_162 + 'Q'
    if E162[n] > E162[n-1]:
        sc_162 = sc_162 + 'E'
    if Z162[n] > Z162[n-1]:
        sc_162 = sc_162 + 'Z'
    if T162[n] > T162[n-1]:
        sc_162 = sc_162 + 'T'
    if len(sc_162) > 0:
        st_162.append(sc_162)
    else:
        STC_162.append(st_162)
        st_162 = []
    n = n + 1


# record the number of empty spatio-temporal sequences
nSTCempty_162 = STC_162.count([])
# keep only non-empty spatio-temporal sequences
STC_162 = [ st_162 for st_162 in STC_162 if len(st_162) > 0 ]

# record the length of the longest spatio-temporal sequence
lmin_162 = len(C162)
lmax_162 = 0
for st_162 in STC_162:
    if len(st_162) < lmin_162:
        lmin_162 = len(st_162)
    if len(st_162) > lmax_162:
        lmax_162 = len(st_162)
        
# restrict spatio-temporal sequences to length-5
STC5_162 = [ st_162 for st_162 in STC_162 if len(st_162) <= 5 ]

# unpack spatio-temporal sequences to temporal sequences
TC5_162 = []
for n in range(len(STC5_162)):
    tc_162 = [ u for u in STC5_162[n][0] ]
    for k in range(1, len(STC5_162[n])):
        tc_162 = [ u+v for u in tc_162 for v in STC5_162[n][k] ]
    TC5_162 = TC5_162 + tc_162
setTC5_162 = list(set(TC5_162))

# get empirical frequencies
fTC5_162 = [ TC5_162.count(setTC5_162[l]) for l in range(len(setTC5_162)) ]

# statistical testing against null model
np.random.seed(0)
fTC5null_162 = [ [] for l in range(len(setTC5_162)) ]
for s in range(162):
    STC5null_162 = [ list(st_162) for st_162 in STC5_162]
    for st_162 in STC5null_162:
        np.random.shuffle(st_162)
    TC5null_162= []
    for n in range(len(STC5null_162)):
        tc_162 = [ u for u in STC5null_162[n][0] ]
        for k in range(1, len(STC5null_162[n])):
            tc_162 = [ u+v for u in tc_162 for v in STC5null_162[n][k] ]
        TC5null_162= TC5null_162+ tc_162
    for l in range(len(setTC5_162)):
        if TC5null_162.count(setTC5_162[l]) > 0:
            fTC5null_162[l].append(TC5null_162.count(setTC5_162[l]))
        else:
            fTC5null_162[l].append(0)
```

```python
print("List of Statistically Significant Sequences at α = 0.05")

#statistical significance testing at p < 0.05 level
for l in range(len(setTC5_162)):
    if sum(freq >= fTC5_162[l] for freq in fTC5null_162[l]) < 5:
        print(setTC5_162[l], fTC5_162[l], sum(freq >= fTC5_162[l] for freq in fTC5null_162[l]))
```

```python
print("List of Statistically Significant Sequences at α = 0.01")

#statistical significance testing at p < 0.01 level
for l in range(len(setTC5_162)):
    if sum(freq >= fTC5_162[l] for freq in fTC5null_162[l]) < 1:
        print(setTC5_162[l], fTC5_162[l], sum(freq >= fTC5_162[l] for freq in fTC5null_162[l]))
```
