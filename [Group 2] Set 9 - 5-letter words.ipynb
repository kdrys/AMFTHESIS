# 121 (L) GTCAP - (E) ALI - (V) TEL - (ε) URC - (W) PGOLD

```python
#extract the overlapping dates

start_121 = max(min(Date_GTCAP), min(Date_ALI), min(Date_TEL), min(Date_URC), min(Date_PGOLD))
end_121 = min(max(Date_GTCAP), max(Date_ALI), max(Date_TEL), max(Date_URC), max(Date_PGOLD))

#shorten the closing prices
GTCAP_121_1 = Date_GTCAP.index(start_121)
GTCAP_121_2 = Date_GTCAP.index(end_121)

ALI_121_1 = Date_ALI.index(start_121)
ALI_121_2 = Date_ALI.index(end_121)

TEL_121_1 = Date_TEL.index(start_121)
TEL_121_2 = Date_TEL.index(end_121)

URC_121_1 = Date_URC.index(start_121)
URC_121_2 = Date_URC.index(end_121)

PGOLD_121_1 = Date_PGOLD.index(start_121)
PGOLD_121_2 = Date_PGOLD.index(end_121)

#save the shortened versions of closing prices
L121 = np.save('GTCAPshortPrices.npy', ClosePrice_GTCAP[GTCAP_121_1:GTCAP_121_2])
E121 = np.save('ALIshortPrices.npy', ClosePrice_ALI[ALI_121_1:ALI_121_2])
V121 = np.save('TELshortPrices.npy', ClosePrice_TEL[TEL_121_1:TEL_121_2])
ε121 = np.save('URCshortPrices.npy', ClosePrice_URC[URC_121_1:URC_121_2])
W121 = np.save('PGOLDshortPrices.npy', ClosePrice_PGOLD[PGOLD_121_1:PGOLD_121_2])

# import modules
import numpy as np
# import prices
L121 = np.load('GTCAPshortPrices.npy', allow_pickle=True).tolist()
E121 = np.load('ALIshortPrices.npy', allow_pickle=True).tolist()
V121 = np.load('TELshortPrices.npy', allow_pickle=True).tolist()
ε121 = np.load('URCshortPrices.npy', allow_pickle=True).tolist()
W121 = np.load('PGOLDshortPrices.npy', allow_pickle=True).tolist()
```

```python
# extract spatio-temporal sequences
STC_121 = []
st_121 = []
n = 1
while n < len(L121):
    sc_121 = ''
    if L121[n] > L121[n-1]:
        sc_121 = sc_121 + 'L'
    if E121[n] > E121[n-1]:
        sc_121 = sc_121 + 'E'
    if V121[n] > V121[n-1]:
        sc_121 = sc_121 + 'V'
    if ε121[n] > ε121[n-1]:
        sc_121 = sc_121 + 'ε'
    if W121[n] > W121[n-1]:
        sc_121 = sc_121 + 'W'
    if len(sc_121) > 0:
        st_121.append(sc_121)
    else:
        STC_121.append(st_121)
        st_121 = []
    n = n + 1


# record the number of empty spatio-temporal sequences
nSTCempty_121 = STC_121.count([])
# keep only non-empty spatio-temporal sequences
STC_121 = [ st_121 for st_121 in STC_121 if len(st_121) > 0 ]

# record the length of the longest spatio-temporal sequence
lmin_121 = len(L121)
lmax_121 = 0
for st_121 in STC_121:
    if len(st_121) < lmin_121:
        lmin_121 = len(st_121)
    if len(st_121) > lmax_121:
        lmax_121 = len(st_121)
        
# restrict spatio-temporal sequences to length-5
STC5_121 = [ st_121 for st_121 in STC_121 if len(st_121) <= 5 ]

# unpack spatio-temporal sequences to temporal sequences
TC5_121 = []
for n in range(len(STC5_121)):
    tc_121 = [ u for u in STC5_121[n][0] ]
    for k in range(1, len(STC5_121[n])):
        tc_121 = [ u+v for u in tc_121 for v in STC5_121[n][k] ]
    TC5_121 = TC5_121 + tc_121
setTC5_121 = list(set(TC5_121))

# get empirical frequencies
fTC5_121 = [ TC5_121.count(setTC5_121[l]) for l in range(len(setTC5_121)) ]

# statistical testing against null model
np.random.seed(0)
fTC5null_121 = [ [] for l in range(len(setTC5_121)) ]
for s in range(121):
    STC5null_121 = [ list(st_121) for st_121 in STC5_121]
    for st_121 in STC5null_121:
        np.random.shuffle(st_121)
    TC5null_121= []
    for n in range(len(STC5null_121)):
        tc_121 = [ u for u in STC5null_121[n][0] ]
        for k in range(1, len(STC5null_121[n])):
            tc_121 = [ u+v for u in tc_121 for v in STC5null_121[n][k] ]
        TC5null_121= TC5null_121+ tc_121
    for l in range(len(setTC5_121)):
        if TC5null_121.count(setTC5_121[l]) > 0:
            fTC5null_121[l].append(TC5null_121.count(setTC5_121[l]))
        else:
            fTC5null_121[l].append(0)
```

```python
print("List of Statistically Significant Sequences at α = 0.05")

#statistical significance testing at p < 0.05 level
for l in range(len(setTC5_121)):
    if sum(freq >= fTC5_121[l] for freq in fTC5null_121[l]) < 5:
        print(setTC5_121[l], fTC5_121[l], sum(freq >= fTC5_121[l] for freq in fTC5null_121[l]))
```

```python
print("List of Statistically Significant Sequences at α = 0.01")

#statistical significance testing at p < 0.01 level
for l in range(len(setTC5_121)):
    if sum(freq >= fTC5_121[l] for freq in fTC5null_121[l]) < 1:
        print(setTC5_121[l], fTC5_121[l], sum(freq >= fTC5_121[l] for freq in fTC5null_121[l]))
```

```python
# extract spatio-temporal sequences
STC_132 = []
st_132 = []
n = 1
while n < len(C132):
    sc_132 = ''
    if C132[n] > C132[n-1]:
        sc_132 = sc_132 + 'C'
    if β132[n] > β132[n-1]:
        sc_132 = sc_132 + 'β'
    if Q132[n] > Q132[n-1]:
        sc_132 = sc_132 + 'Q'
    if Z132[n] > Z132[n-1]:
        sc_132 = sc_132 + 'Z'
    if T132[n] > T132[n-1]:
        sc_132 = sc_132 + 'T'
    if len(sc_132) > 0:
        st_132.append(sc_132)
    else:
        STC_132.append(st_132)
        st_132 = []
    n = n + 1


# record the number of empty spatio-temporal sequences
nSTCempty_132 = STC_132.count([])
# keep only non-empty spatio-temporal sequences
STC_132 = [ st_132 for st_132 in STC_132 if len(st_132) > 0 ]

# record the length of the longest spatio-temporal sequence
lmin_132 = len(C132)
lmax_132 = 0
for st_132 in STC_132:
    if len(st_132) < lmin_132:
        lmin_132 = len(st_132)
    if len(st_132) > lmax_132:
        lmax_132 = len(st_132)
        
# restrict spatio-temporal sequences to length-5
STC5_132 = [ st_132 for st_132 in STC_132 if len(st_132) <= 5 ]

# unpack spatio-temporal sequences to temporal sequences
TC5_132 = []
for n in range(len(STC5_132)):
    tc_132 = [ u for u in STC5_132[n][0] ]
    for k in range(1, len(STC5_132[n])):
        tc_132 = [ u+v for u in tc_132 for v in STC5_132[n][k] ]
    TC5_132 = TC5_132 + tc_132
setTC5_132 = list(set(TC5_132))

# get empirical frequencies
fTC5_132 = [ TC5_132.count(setTC5_132[l]) for l in range(len(setTC5_132)) ]

# statistical testing against null model
np.random.seed(0)
fTC5null_132 = [ [] for l in range(len(setTC5_132)) ]
for s in range(132):
    STC5null_132 = [ list(st_132) for st_132 in STC5_132]
    for st_132 in STC5null_132:
        np.random.shuffle(st_132)
    TC5null_132= []
    for n in range(len(STC5null_132)):
        tc_132 = [ u for u in STC5null_132[n][0] ]
        for k in range(1, len(STC5null_132[n])):
            tc_132 = [ u+v for u in tc_132 for v in STC5null_132[n][k] ]
        TC5null_132= TC5null_132+ tc_132
    for l in range(len(setTC5_132)):
        if TC5null_132.count(setTC5_132[l]) > 0:
            fTC5null_132[l].append(TC5null_132.count(setTC5_132[l]))
        else:
            fTC5null_132[l].append(0)
```

```python
print("List of Statistically Significant Sequences at α = 0.05")

#statistical significance testing at p < 0.05 level
for l in range(len(setTC5_132)):
    if sum(freq >= fTC5_132[l] for freq in fTC5null_132[l]) < 5:
        print(setTC5_132[l], fTC5_132[l], sum(freq >= fTC5_132[l] for freq in fTC5null_132[l]))
```
