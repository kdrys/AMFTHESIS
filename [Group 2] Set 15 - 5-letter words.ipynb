# 211 (β) SM - (Q) MER - (Z) SECB - (V) TEL - (W) PGOLD

```python
#extract the overlapping dates

start_211 = max(min(Date_SM), min(Date_MER), min(Date_SECB), min(Date_TEL), min(Date_PGOLD))
end_211 = min(max(Date_SM), max(Date_MER), max(Date_SECB), max(Date_TEL), max(Date_PGOLD))

#shorten the closing prices
SM_211_1 = Date_SM.index(start_211)
SM_211_2 = Date_SM.index(end_211)

MER_211_1 = Date_MER.index(start_211)
MER_211_2 = Date_MER.index(end_211)

SECB_211_1 = Date_SECB.index(start_211)
SECB_211_2 = Date_SECB.index(end_211)

TEL_211_1 = Date_TEL.index(start_211)
TEL_211_2 = Date_TEL.index(end_211)

PGOLD_211_1 = Date_PGOLD.index(start_211)
PGOLD_211_2 = Date_PGOLD.index(end_211)

#save the shortened versions of closing prices
β211 = np.save('SMshortPrices.npy', ClosePrice_SM[SM_211_1:SM_211_2])
Q211 = np.save('MERshortPrices.npy', ClosePrice_MER[MER_211_1:MER_211_2])
Z211 = np.save('SECBshortPrices.npy', ClosePrice_SECB[SECB_211_1:SECB_211_2])
V211 = np.save('TELshortPrices.npy', ClosePrice_TEL[TEL_211_1:TEL_211_2])
W211 = np.save('PGOLDshortPrices.npy', ClosePrice_PGOLD[PGOLD_211_1:PGOLD_211_2])

# import modules
import numpy as np
# import prices
β211 = np.load('SMshortPrices.npy', allow_pickle=True).tolist()
Q211 = np.load('MERshortPrices.npy', allow_pickle=True).tolist()
Z211 = np.load('SECBshortPrices.npy', allow_pickle=True).tolist()
V211 = np.load('TELshortPrices.npy', allow_pickle=True).tolist()
W211 = np.load('PGOLDshortPrices.npy', allow_pickle=True).tolist()
```

```python
# extract spatio-temporal sequences
STC_211 = []
st_211 = []
n = 1
while n < len(β211):
    sc_211 = ''
    if β211[n] > β211[n-1]:
        sc_211 = sc_211 + 'β'
    if Q211[n] > Q211[n-1]:
        sc_211 = sc_211 + 'Q'
    if Z211[n] > Z211[n-1]:
        sc_211 = sc_211 + 'Z'
    if V211[n] > V211[n-1]:
        sc_211 = sc_211 + 'V'
    if W211[n] > W211[n-1]:
        sc_211 = sc_211 + 'W'
    if len(sc_211) > 0:
        st_211.append(sc_211)
    else:
        STC_211.append(st_211)
        st_211 = []
    n = n + 1


# record the number of empty spatio-temporal sequences
nSTCempty_211 = STC_211.count([])
# keep only non-empty spatio-temporal sequences
STC_211 = [ st_211 for st_211 in STC_211 if len(st_211) > 0 ]

# record the length of the longest spatio-temporal sequence
lmin_211 = len(β211)
lmax_211 = 0
for st_211 in STC_211:
    if len(st_211) < lmin_211:
        lmin_211 = len(st_211)
    if len(st_211) > lmax_211:
        lmax_211 = len(st_211)
        
# restrict spatio-temporal sequences to length-5
STC5_211 = [ st_211 for st_211 in STC_211 if len(st_211) <= 5 ]

# unpack spatio-temporal sequences to temporal sequences
TC5_211 = []
for n in range(len(STC5_211)):
    tc_211 = [ u for u in STC5_211[n][0] ]
    for k in range(1, len(STC5_211[n])):
        tc_211 = [ u+v for u in tc_211 for v in STC5_211[n][k] ]
    TC5_211 = TC5_211 + tc_211
setTC5_211 = list(set(TC5_211))

# get empirical frequencies
fTC5_211 = [ TC5_211.count(setTC5_211[l]) for l in range(len(setTC5_211)) ]

# statistical testing against null model
np.random.seed(0)
fTC5null_211 = [ [] for l in range(len(setTC5_211)) ]
for s in range(211):
    STC5null_211 = [ list(st_211) for st_211 in STC5_211]
    for st_211 in STC5null_211:
        np.random.shuffle(st_211)
    TC5null_211= []
    for n in range(len(STC5null_211)):
        tc_211 = [ u for u in STC5null_211[n][0] ]
        for k in range(1, len(STC5null_211[n])):
            tc_211 = [ u+v for u in tc_211 for v in STC5null_211[n][k] ]
        TC5null_211= TC5null_211+ tc_211
    for l in range(len(setTC5_211)):
        if TC5null_211.count(setTC5_211[l]) > 0:
            fTC5null_211[l].append(TC5null_211.count(setTC5_211[l]))
        else:
            fTC5null_211[l].append(0)
```

```python
print("List of Statistically Significant Sequences at α = 0.05")

#statistical significance testing at p < 0.05 level
for l in range(len(setTC5_211)):
    if sum(freq >= fTC5_211[l] for freq in fTC5null_211[l]) < 5:
        print(setTC5_211[l], fTC5_211[l], sum(freq >= fTC5_211[l] for freq in fTC5null_211[l]))
```

```python
print("List of Statistically Significant Sequences at α = 0.01")

#statistical significance testing at p < 0.01 level
for l in range(len(setTC5_211)):
    if sum(freq >= fTC5_211[l] for freq in fTC5null_211[l]) < 1:
        print(setTC5_211[l], fTC5_211[l], sum(freq >= fTC5_211[l] for freq in fTC5null_211[l]))
```

```python
# extract spatio-temporal sequences
STC_218 = []
st_218 = []
n = 1
while n < len(β218):
    sc_218 = ''
    if β218[n] > β218[n-1]:
        sc_218 = sc_218 + 'β'
    if E218[n] > E218[n-1]:
        sc_218 = sc_218 + 'E'
    if Z218[n] > Z218[n-1]:
        sc_218 = sc_218 + 'Z'
    if T218[n] > T218[n-1]:
        sc_218 = sc_218 + 'T'
    if ε218[n] > ε218[n-1]:
        sc_218 = sc_218 + 'ε'
    if len(sc_218) > 0:
        st_218.append(sc_218)
    else:
        STC_218.append(st_218)
        st_218 = []
    n = n + 1


# record the number of empty spatio-temporal sequences
nSTCempty_218 = STC_218.count([])
# keep only non-empty spatio-temporal sequences
STC_218 = [ st_218 for st_218 in STC_218 if len(st_218) > 0 ]

# record the length of the longest spatio-temporal sequence
lmin_218 = len(β218)
lmax_218 = 0
for st_218 in STC_218:
    if len(st_218) < lmin_218:
        lmin_218 = len(st_218)
    if len(st_218) > lmax_218:
        lmax_218 = len(st_218)
        
# restrict spatio-temporal sequences to length-5
STC5_218 = [ st_218 for st_218 in STC_218 if len(st_218) <= 5 ]

# unpack spatio-temporal sequences to temporal sequences
TC5_218 = []
for n in range(len(STC5_218)):
    tc_218 = [ u for u in STC5_218[n][0] ]
    for k in range(1, len(STC5_218[n])):
        tc_218 = [ u+v for u in tc_218 for v in STC5_218[n][k] ]
    TC5_218 = TC5_218 + tc_218
setTC5_218 = list(set(TC5_218))

# get empirical frequencies
fTC5_218 = [ TC5_218.count(setTC5_218[l]) for l in range(len(setTC5_218)) ]

# statistical testing against null model
np.random.seed(0)
fTC5null_218 = [ [] for l in range(len(setTC5_218)) ]
for s in range(218):
    STC5null_218 = [ list(st_218) for st_218 in STC5_218]
    for st_218 in STC5null_218:
        np.random.shuffle(st_218)
    TC5null_218= []
    for n in range(len(STC5null_218)):
        tc_218 = [ u for u in STC5null_218[n][0] ]
        for k in range(1, len(STC5null_218[n])):
            tc_218 = [ u+v for u in tc_218 for v in STC5null_218[n][k] ]
        TC5null_218= TC5null_218+ tc_218
    for l in range(len(setTC5_218)):
        if TC5null_218.count(setTC5_218[l]) > 0:
            fTC5null_218[l].append(TC5null_218.count(setTC5_218[l]))
        else:
            fTC5null_218[l].append(0)
```

```python
# extract spatio-temporal sequences
STC_222 = []
st_222 = []
n = 1
while n < len(β222):
    sc_222 = ''
    if β222[n] > β222[n-1]:
        sc_222 = sc_222 + 'β'
    if E222[n] > E222[n-1]:
        sc_222 = sc_222 + 'E'
    if Z222[n] > Z222[n-1]:
        sc_222 = sc_222 + 'Z'
    if ε222[n] > ε222[n-1]:
        sc_222 = sc_222 + 'ε'
    if W222[n] > W222[n-1]:
        sc_222 = sc_222 + 'W'
    if len(sc_222) > 0:
        st_222.append(sc_222)
    else:
        STC_222.append(st_222)
        st_222 = []
    n = n + 1


# record the number of empty spatio-temporal sequences
nSTCempty_222 = STC_222.count([])
# keep only non-empty spatio-temporal sequences
STC_222 = [ st_222 for st_222 in STC_222 if len(st_222) > 0 ]

# record the length of the longest spatio-temporal sequence
lmin_222 = len(β222)
lmax_222 = 0
for st_222 in STC_222:
    if len(st_222) < lmin_222:
        lmin_222 = len(st_222)
    if len(st_222) > lmax_222:
        lmax_222 = len(st_222)
        
# restrict spatio-temporal sequences to length-5
STC5_222 = [ st_222 for st_222 in STC_222 if len(st_222) <= 5 ]

# unpack spatio-temporal sequences to temporal sequences
TC5_222 = []
for n in range(len(STC5_222)):
    tc_222 = [ u for u in STC5_222[n][0] ]
    for k in range(1, len(STC5_222[n])):
        tc_222 = [ u+v for u in tc_222 for v in STC5_222[n][k] ]
    TC5_222 = TC5_222 + tc_222
setTC5_222 = list(set(TC5_222))

# get empirical frequencies
fTC5_222 = [ TC5_222.count(setTC5_222[l]) for l in range(len(setTC5_222)) ]

# statistical testing against null model
np.random.seed(0)
fTC5null_222 = [ [] for l in range(len(setTC5_222)) ]
for s in range(222):
    STC5null_222 = [ list(st_222) for st_222 in STC5_222]
    for st_222 in STC5null_222:
        np.random.shuffle(st_222)
    TC5null_222= []
    for n in range(len(STC5null_222)):
        tc_222 = [ u for u in STC5null_222[n][0] ]
        for k in range(1, len(STC5null_222[n])):
            tc_222 = [ u+v for u in tc_222 for v in STC5null_222[n][k] ]
        TC5null_222= TC5null_222+ tc_222
    for l in range(len(setTC5_222)):
        if TC5null_222.count(setTC5_222[l]) > 0:
            fTC5null_222[l].append(TC5null_222.count(setTC5_222[l]))
        else:
            fTC5null_222[l].append(0)
```

```python
print("List of Statistically Significant Sequences at α = 0.05")

#statistical significance testing at p < 0.05 level
for l in range(len(setTC5_222)):
    if sum(freq >= fTC5_222[l] for freq in fTC5null_222[l]) < 5:
        print(setTC5_222[l], fTC5_222[l], sum(freq >= fTC5_222[l] for freq in fTC5null_222[l]))
```

```python
print("List of Statistically Significant Sequences at α = 0.01")

#statistical significance testing at p < 0.01 level
for l in range(len(setTC5_222)):
    if sum(freq >= fTC5_222[l] for freq in fTC5null_222[l]) < 1:
        print(setTC5_222[l], fTC5_222[l], sum(freq >= fTC5_222[l] for freq in fTC5null_222[l]))
```
