```python
# import modules
import numpy as np
# import prices
B1 = np.load('APshortPrices.npy', allow_pickle=True).tolist() 
D1 = np.load('ACshortPrices.npy', allow_pickle=True).tolist() 
F1 = np.load('BPIshortPrices.npy', allow_pickle=True).tolist() 
H1 = np.load('BLOOMshortPrices.npy', allow_pickle=True).tolist() 
J1 = np.load('FGENshortPrices.npy', allow_pickle=True).tolist()
M1 = np.load('ICTshortPrices.npy', allow_pickle=True).tolist()
N1 = np.load('JGSshortPrices.npy', allow_pickle=True).tolist()
R1 = np.load('MEGshortPrices.npy', allow_pickle=True).tolist()
Y1 = np.load('SMCshortPrices.npy', allow_pickle=True).tolist()
α1 = np.load('SCCshortPrices.npy', allow_pickle=True).tolist()
```

```python
# extract spatio-temporal sequences
STC_1 = []
st_1 = []
n = 1
while n < len(B1):
    sc_1 = ''
    if B1[n] > B1[n-1]:
        sc_1 = sc_1 + 'B'
    if D1[n] > D1[n-1]:
        sc_1 = sc_1 + 'D'
    if F1[n] > F1[n-1]:
        sc_1 = sc_1 + 'F'
    if H1[n] > H1[n-1]:
        sc_1 = sc_1 + 'H'
    if J1[n] > J1[n-1]: 
        sc_1 = sc_1 + 'J'
    if M1[n] > M1[n-1]:
        sc_1 = sc_1 + 'M'
    if N1[n] > N1[n-1]:
        sc_1 = sc_1 + 'N'
    if R1[n] > R1[n-1]:
        sc_1 = sc_1 + 'R'
    if Y1[n] > Y1[n-1]:
        sc_1 = sc_1 + 'Y'
    if α1[n] > α1[n-1]: 
        sc_1 = sc_1 + 'α'
    if len(sc_1) > 0:
        st_1.append(sc_1)
    else:
        STC_1.append(st_1)
        st_1 = []
    n = n + 1
```

```python
# record the number of empty spatio-temporal sequences
nSTCempty_1 = STC_1.count([])
# keep only non-empty spatio-temporal sequences
STC_1 = [ st_1 for st_1 in STC_1 if len(st_1) > 0 ]

# record the length of the longest spatio-temporal sequence
lmin_1 = len(B1)
lmax_1 = 0
for st_1 in STC_1:
    if len(st_1) < lmin_1:
        lmin_1 = len(st_1)
    if len(st_1) > lmax_1:
        lmax_1 = len(st_1)
        
# restrict spatio-temporal sequences to length-5
STC5_1 = [ st_1 for st_1 in STC_1 if len(st_1) <= 5 ]

# unpack spatio-temporal sequences to temporal sequences
TC5_1 = []
for n in range(len(STC5_1)):
    tc_1 = [ u for u in STC5_1[n][0] ]
    for k in range(1, len(STC5_1[n])):
        tc_1 = [ u+v for u in tc_1 for v in STC5_1[n][k] ]
    TC5_1 = TC5_1 + tc_1
setTC5_1 = list(set(TC5_1))

# get empirical frequencies
fTC5_1 = [ TC5_1.count(setTC5_1[l]) for l in range(len(setTC5_1)) ]

# statistical testing against null model
np.random.seed(0)
fTC5null_1 = [ [] for l in range(len(setTC5_1)) ]
for s in range(100):
    STC5null_1 = [ list(st_1) for st_1 in STC5_1]
    for st_1 in STC5null_1:
        np.random.shuffle(st_1)
    TC5null_1 = []
    for n in range(len(STC5null_1)):
        tc_1 = [ u for u in STC5null_1[n][0] ]
        for k in range(1, len(STC5null_1[n])):
            tc_1 = [ u+v for u in tc_1 for v in STC5null_1[n][k] ]
        TC5null_1 = TC5null_1 + tc_1
    for l in range(len(setTC5_1)):
        if TC5null_1.count(setTC5_1[l]) > 0:
            fTC5null_1[l].append(TC5null_1.count(setTC5_1[l]))
        else:
            fTC5null_1[l].append(0)
```
