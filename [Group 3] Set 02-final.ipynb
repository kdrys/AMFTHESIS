```python
#shorten the closing prices
AP_16_1 = Date_AP.index(start_16)
AP_16_2 = Date_AP.index(end_16)

AC_16_1 = Date_AC.index(start_16)
AC_16_2 = Date_AC.index(end_16)

BPI_16_1 = Date_BPI.index(start_16)
BPI_16_2 = Date_BPI.index(end_16)

JGS_16_1 = Date_JGS.index(start_16)
JGS_16_2 = Date_JGS.index(end_16)

MEG_16_1 = Date_MEG.index(start_16)
MEG_16_2 = Date_MEG.index(end_16)
```

```python
#save the shortened versions of closing prices
B16 = np.save('APshortPrices.npy', ClosePrice_AP[AP_16_1:AP_16_2])
D16 = np.save('ACshortPrices.npy', ClosePrice_AC[AC_16_1:AC_16_2])
F16 = np.save('BPIshortPrices.npy', ClosePrice_BPI[BPI_16_1:BPI_16_2])
N16 = np.save('JGSshortPrices.npy', ClosePrice_JGS[JGS_16_1:JGS_16_2])
R16 = np.save('MEGshortPrices.npy', ClosePrice_MEG[MEG_16_1:MEG_16_2])
```

```python
# import modules
import numpy as np
# import prices

B16 = np.load('APshortPrices.npy', allow_pickle=True).tolist() 
D16 = np.load('ACshortPrices.npy', allow_pickle=True).tolist() 
F16 = np.load('BPIshortPrices.npy', allow_pickle=True).tolist() 
N16 = np.load('JGSshortPrices.npy', allow_pickle=True).tolist() 
R16 = np.load('MEGshortPrices.npy', allow_pickle=True).tolist()
```

```python
# extract spatio-temporal sequences
STC_16 = []
st_16 = []
n = 1
while n < len(B16):
    sc_16 = ''
    if B16[n] > B16[n-1]:
        sc_16 = sc_16 + 'B'
    if D16[n] > D16[n-1]:
        sc_16 = sc_16 + 'D'
    if F16[n] > F16[n-1]:
        sc_16 = sc_16 + 'F'
    if N16[n] > N16[n-1]:
        sc_16 = sc_16 + 'N'
    if R16[n] > R16[n-1]:
        sc_16 = sc_16 + 'Î±'
    if len(sc_16) > 0:
        st_16.append(sc_16)
    else:
        STC_16.append(st_16)
        st_16 = []
    n = n + 1
```

```python
# record the number of empty spatio-temporal sequences
nSTCempty_16 = STC_16.count([])
# keep only non-empty spatio-temporal sequences
STC_16 = [ st_16 for st_16 in STC_16 if len(st_16) > 0 ]

# record the length of the longest spatio-temporal sequence
lmin_16 = len(B16)
lmax_16 = 0
for st_16 in STC_16:
    if len(st_16) < lmin_16:
        lmin_16 = len(st_16)
    if len(st_16) > lmax_16:
        lmax_16 = len(st_16)
        
# restrict spatio-temporal sequences to length-5
STC5_16 = [ st_16 for st_16 in STC_16 if len(st_16) <= 5 ]

# unpack spatio-temporal sequences to temporal sequences
TC5_16 = []
for n in range(len(STC5_16)):
    tc_16 = [ u for u in STC5_16[n][0] ]
    for k in range(1, len(STC5_16[n])):
        tc_16 = [ u+v for u in tc_16 for v in STC5_16[n][k] ]
    TC5_16 = TC5_16 + tc_16
setTC5_16 = list(set(TC5_16))

# get empirical frequencies
fTC5_16 = [ TC5_16.count(setTC5_16[l]) for l in range(len(setTC5_16)) ]

# statistical testing against null model
np.random.seed(0)
fTC5null_16 = [ [] for l in range(len(setTC5_16)) ]
for s in range(100):
    STC5null_16 = [ list(st_16) for st_16 in STC5_16]
    for st_16 in STC5null_16:
        np.random.shuffle(st_16)
    TC5null_16= []
    for n in range(len(STC5null_16)):
        tc_16 = [ u for u in STC5null_16[n][0] ]
        for k in range(1, len(STC5null_16[n])):
            tc_16 = [ u+v for u in tc_16 for v in STC5null_16[n][k] ]
        TC5null_16= TC5null_16+ tc_16
    for l in range(len(setTC5_16)):
        if TC5null_16.count(setTC5_16[l]) > 0:
            fTC5null_16[l].append(TC5null_16.count(setTC5_16[l]))
        else:
            fTC5null_16[l].append(0)
```

```python
print("List of Statistically Significant Sequences at R = 0.05")

#statistical significance testing at p < 0.05 level
for l in range(len(setTC5_16)):
     if sum(freq >= fTC5_16[l] for freq in fTC5null_16[l]) < 5:
        print(setTC5_16[l], fTC5_16[l], sum(freq >= fTC5_16[l] for freq in fTC5null_16[l]))
#         print(setTC5_16[l])
```

```python
print("List of Statistically Significant Sequences at R = 0.01")

#statistical significance testing at p < 0.01 level
for l in range(len(setTC5_16)):
     if sum(freq >= fTC5_16[l] for freq in fTC5null_16[l]) < 1:
        print(setTC5_16[l], fTC5_16[l], sum(freq >= fTC5_16[l] for freq in fTC5null_16[l]))
#         print(setTC5_16[l])
```
