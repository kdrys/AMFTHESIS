```python
# import modules
import numpy as np
# import prices

D151 = np.load('ACshortPrices.npy', allow_pickle=True).tolist() 
F151 = np.load('BPIshortPrices.npy', allow_pickle=True).tolist() 
J151 = np.load('FGENshortPrices.npy', allow_pickle=True).tolist() 
Y151 = np.load('SMCshortPrices.npy', allow_pickle=True).tolist() 
α151 = np.load('SCCshortPrices.npy', allow_pickle=True).tolist()
```

```python
# extract spatio-temporal sequences
STC_151 = []
st_151 = []
n = 1
while n < len(D151):
    sc_151 = ''
    if D151[n] > D151[n-1]:
        sc_151 = sc_151 + 'D'
    if F151[n] > F151[n-1]:
        sc_151 = sc_151 + 'F'
    if J151[n] > J151[n-1]:
        sc_151 = sc_151 + 'J'
    if Y151[n] > Y151[n-1]:
        sc_151 = sc_151 + 'R'
    if α151[n] > α151[n-1]:
        sc_151 = sc_151 + 'α'
    if len(sc_151) > 0:
        st_151.append(sc_151)
    else:
        STC_151.append(st_151)
        st_151 = []
    n = n + 1
```

```python
# record the number of empty spatio-temporal sequences
nSTCempty_151 = STC_151.count([])
# keep only non-empty spatio-temporal sequences
STC_151 = [ st_151 for st_151 in STC_151 if len(st_151) > 0 ]

# record the length of the longest spatio-temporal sequence
lmin_151 = len(D151)
lmax_151 = 0
for st_151 in STC_151:
    if len(st_151) < lmin_151:
        lmin_151 = len(st_151)
    if len(st_151) > lmax_151:
        lmax_151 = len(st_151)
        
# restrSMC spatio-temporal sequences to length-5
STC5_151 = [ st_151 for st_151 in STC_151 if len(st_151) <= 5 ]

# unpack spatio-temporal sequences to temporal sequences
TC5_151 = []
for n in range(len(STC5_151)):
    tc_151 = [ u for u in STC5_151[n][0] ]
    for k in range(1, len(STC5_151[n])):
        tc_151 = [ u+v for u in tc_151 for v in STC5_151[n][k] ]
    TC5_151 = TC5_151 + tc_151
setTC5_151 = list(set(TC5_151))

# get empirical frequencies
fTC5_151 = [ TC5_151.count(setTC5_151[l]) for l in range(len(setTC5_151)) ]

# statistical testing against null model
np.random.seed(0)
fTC5null_151 = [ [] for l in range(len(setTC5_151)) ]
for s in range(100):
    STC5null_151 = [ list(st_151) for st_151 in STC5_151]
    for st_151 in STC5null_151:
        np.random.shuffle(st_151)
    TC5null_151= []
    for n in range(len(STC5null_151)):
        tc_151 = [ u for u in STC5null_151[n][0] ]
        for k in range(1, len(STC5null_151[n])):
            tc_151 = [ u+v for u in tc_151 for v in STC5null_151[n][k] ]
        TC5null_151= TC5null_151+ tc_151
    for l in range(len(setTC5_151)):
        if TC5null_151.count(setTC5_151[l]) > 0:
            fTC5null_151[l].append(TC5null_151.count(setTC5_151[l]))
        else:
            fTC5null_151[l].append(0)
```

```python
print("List of Statistically Significant Sequences at R = 0.05")

#statistical significance testing at p < 0.05 level
for l in range(len(setTC5_151)):
     if sum(freq >= fTC5_151[l] for freq in fTC5null_151[l]) < 5:
#         print(setTC5_151[l], fTC5_151[l], sum(freq >= fTC5_151[l] for freq in fTC5null_151[l]))
        print(setTC5_151[l])
```

```python
print("List of Statistically Significant Sequences at R = 0.01")

#statistical significance testing at p < 0.01 level
for l in range(len(setTC5_151)):
     if sum(freq >= fTC5_151[l] for freq in fTC5null_151[l]) < 1:
#         print(setTC5_151[l], fTC5_151[l], sum(freq >= fTC5_151[l] for freq in fTC5null_151[l]))
        print(setTC5_151[l])
```
