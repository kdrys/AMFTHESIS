# 196 (C) AGI - (T) MBT - (V) TEL - (ε) URC - (W) PGOLD 

```python
#extract the overlapping dates

start_196 = max(min(Date_AGI), min(Date_MBT), min(Date_TEL), min(Date_URC), min(Date_PGOLD))
end_196 = min(max(Date_AGI), max(Date_MBT), max(Date_TEL), max(Date_URC), max(Date_PGOLD))

#shorten the closing prices
AGI_196_1 = Date_AGI.index(start_196)
AGI_196_2 = Date_AGI.index(end_196)

MBT_196_1 = Date_MBT.index(start_196)
MBT_196_2 = Date_MBT.index(end_196)

TEL_196_1 = Date_TEL.index(start_196)
TEL_196_2 = Date_TEL.index(end_196)

URC_196_1 = Date_URC.index(start_196)
URC_196_2 = Date_URC.index(end_196)

PGOLD_196_1 = Date_PGOLD.index(start_196)
PGOLD_196_2 = Date_PGOLD.index(end_196)

#save the shortened versions of closing prices
C196 = np.save('AGIshortPrices.npy', ClosePrice_AGI[AGI_196_1:AGI_196_2])
T196 = np.save('MBTshortPrices.npy', ClosePrice_MBT[MBT_196_1:MBT_196_2])
V196 = np.save('TELshortPrices.npy', ClosePrice_TEL[TEL_196_1:TEL_196_2])
ε196 = np.save('URCshortPrices.npy', ClosePrice_URC[URC_196_1:URC_196_2])
W196 = np.save('PGOLDshortPrices.npy', ClosePrice_PGOLD[PGOLD_196_1:PGOLD_196_2])

# import modules
import numpy as np
# import prices
C196 = np.load('AGIshortPrices.npy', allow_pickle=True).tolist()
T196 = np.load('MBTshortPrices.npy', allow_pickle=True).tolist()
V196 = np.load('TELshortPrices.npy', allow_pickle=True).tolist()
ε196 = np.load('URCshortPrices.npy', allow_pickle=True).tolist()
W196 = np.load('PGOLDshortPrices.npy', allow_pickle=True).tolist()
```

```python
# extract spatio-temporal sequences
STC_196 = []
st_196 = []
n = 1
while n < len(C196):
    sc_196 = ''
    if C196[n] > C196[n-1]:
        sc_196 = sc_196 + 'C'
    if T196[n] > T196[n-1]:
        sc_196 = sc_196 + 'T'
    if V196[n] > V196[n-1]:
        sc_196 = sc_196 + 'V'
    if ε196[n] > ε196[n-1]:
        sc_196 = sc_196 + 'ε'
    if W196[n] > W196[n-1]:
        sc_196 = sc_196 + 'W'
    if len(sc_196) > 0:
        st_196.append(sc_196)
    else:
        STC_196.append(st_196)
        st_196 = []
    n = n + 1


# record the number of empty spatio-temporal sequences
nSTCempty_196 = STC_196.count([])
# keep only non-empty spatio-temporal sequences
STC_196 = [ st_196 for st_196 in STC_196 if len(st_196) > 0 ]

# record the length of the longest spatio-temporal sequence
lmin_196 = len(C196)
lmax_196 = 0
for st_196 in STC_196:
    if len(st_196) < lmin_196:
        lmin_196 = len(st_196)
    if len(st_196) > lmax_196:
        lmax_196 = len(st_196)
        
# restrict spatio-temporal sequences to length-5
STC5_196 = [ st_196 for st_196 in STC_196 if len(st_196) <= 5 ]

# unpack spatio-temporal sequences to temporal sequences
TC5_196 = []
for n in range(len(STC5_196)):
    tc_196 = [ u for u in STC5_196[n][0] ]
    for k in range(1, len(STC5_196[n])):
        tc_196 = [ u+v for u in tc_196 for v in STC5_196[n][k] ]
    TC5_196 = TC5_196 + tc_196
setTC5_196 = list(set(TC5_196))

# get empirical frequencies
fTC5_196 = [ TC5_196.count(setTC5_196[l]) for l in range(len(setTC5_196)) ]

# statistical testing against null model
np.random.seed(0)
fTC5null_196 = [ [] for l in range(len(setTC5_196)) ]
for s in range(196):
    STC5null_196 = [ list(st_196) for st_196 in STC5_196]
    for st_196 in STC5null_196:
        np.random.shuffle(st_196)
    TC5null_196= []
    for n in range(len(STC5null_196)):
        tc_196 = [ u for u in STC5null_196[n][0] ]
        for k in range(1, len(STC5null_196[n])):
            tc_196 = [ u+v for u in tc_196 for v in STC5null_196[n][k] ]
        TC5null_196= TC5null_196+ tc_196
    for l in range(len(setTC5_196)):
        if TC5null_196.count(setTC5_196[l]) > 0:
            fTC5null_196[l].append(TC5null_196.count(setTC5_196[l]))
        else:
            fTC5null_196[l].append(0)
```

```python
print("List of Statistically Significant Sequences at α = 0.05")

#statistical significance testing at p < 0.05 level
for l in range(len(setTC5_196)):
    if sum(freq >= fTC5_196[l] for freq in fTC5null_196[l]) < 5:
        print(setTC5_196[l], fTC5_196[l], sum(freq >= fTC5_196[l] for freq in fTC5null_196[l]))
```

```python
print("List of Statistically Significant Sequences at α = 0.01")

#statistical significance testing at p < 0.01 level
for l in range(len(setTC5_196)):
    if sum(freq >= fTC5_196[l] for freq in fTC5null_196[l]) < 1:
        print(setTC5_196[l], fTC5_196[l], sum(freq >= fTC5_196[l] for freq in fTC5null_196[l]))
```

```python
# extract spatio-temporal sequences
STC_199 = []
st_199 = []
n = 1
while n < len(β199):
    sc_199 = ''
    if β199[n] > β199[n-1]:
        sc_199 = sc_199 + 'β'
    if Q199[n] > Q199[n-1]:
        sc_199 = sc_199 + 'Q'
    if E199[n] > E199[n-1]:
        sc_199 = sc_199 + 'E'
    if Z199[n] > Z199[n-1]:
        sc_199 = sc_199 + 'Z'
    if ε199[n] > ε199[n-1]:
        sc_199 = sc_199 + 'ε'
    if len(sc_199) > 0:
        st_199.append(sc_199)
    else:
        STC_199.append(st_199)
        st_199 = []
    n = n + 1


# record the number of empty spatio-temporal sequences
nSTCempty_199 = STC_199.count([])
# keep only non-empty spatio-temporal sequences
STC_199 = [ st_199 for st_199 in STC_199 if len(st_199) > 0 ]

# record the length of the longest spatio-temporal sequence
lmin_199 = len(β199)
lmax_199 = 0
for st_199 in STC_199:
    if len(st_199) < lmin_199:
        lmin_199 = len(st_199)
    if len(st_199) > lmax_199:
        lmax_199 = len(st_199)
        
# restrict spatio-temporal sequences to length-5
STC5_199 = [ st_199 for st_199 in STC_199 if len(st_199) <= 5 ]

# unpack spatio-temporal sequences to temporal sequences
TC5_199 = []
for n in range(len(STC5_199)):
    tc_199 = [ u for u in STC5_199[n][0] ]
    for k in range(1, len(STC5_199[n])):
        tc_199 = [ u+v for u in tc_199 for v in STC5_199[n][k] ]
    TC5_199 = TC5_199 + tc_199
setTC5_199 = list(set(TC5_199))

# get empirical frequencies
fTC5_199 = [ TC5_199.count(setTC5_199[l]) for l in range(len(setTC5_199)) ]

# statistical testing against null model
np.random.seed(0)
fTC5null_199 = [ [] for l in range(len(setTC5_199)) ]
for s in range(199):
    STC5null_199 = [ list(st_199) for st_199 in STC5_199]
    for st_199 in STC5null_199:
        np.random.shuffle(st_199)
    TC5null_199= []
    for n in range(len(STC5null_199)):
        tc_199 = [ u for u in STC5null_199[n][0] ]
        for k in range(1, len(STC5null_199[n])):
            tc_199 = [ u+v for u in tc_199 for v in STC5null_199[n][k] ]
        TC5null_199= TC5null_199+ tc_199
    for l in range(len(setTC5_199)):
        if TC5null_199.count(setTC5_199[l]) > 0:
            fTC5null_199[l].append(TC5null_199.count(setTC5_199[l]))
        else:
            fTC5null_199[l].append(0)
```

```python
print("List of Statistically Significant Sequences at α = 0.05")

#statistical significance testing at p < 0.05 level
for l in range(len(setTC5_199)):
    if sum(freq >= fTC5_199[l] for freq in fTC5null_199[l]) < 5:
        print(setTC5_199[l], fTC5_199[l], sum(freq >= fTC5_199[l] for freq in fTC5null_199[l]))
```

```python
# extract spatio-temporal sequences
STC_207 = []
st_207 = []
n = 1
while n < len(β207):
    sc_207 = ''
    if β207[n] > β207[n-1]:
        sc_207 = sc_207 + 'β'
    if Q207[n] > Q207[n-1]:
        sc_207 = sc_207 + 'Q'
    if Z207[n] > Z207[n-1]:
        sc_207 = sc_207 + 'Z'
    if T207[n] > T207[n-1]:
        sc_207 = sc_207 + 'T'
    if V207[n] > V207[n-1]:
        sc_207 = sc_207 + 'V'
    if len(sc_207) > 0:
        st_207.append(sc_207)
    else:
        STC_207.append(st_207)
        st_207 = []
    n = n + 1


# record the number of empty spatio-temporal sequences
nSTCempty_207 = STC_207.count([])
# keep only non-empty spatio-temporal sequences
STC_207 = [ st_207 for st_207 in STC_207 if len(st_207) > 0 ]

# record the length of the longest spatio-temporal sequence
lmin_207 = len(β207)
lmax_207 = 0
for st_207 in STC_207:
    if len(st_207) < lmin_207:
        lmin_207 = len(st_207)
    if len(st_207) > lmax_207:
        lmax_207 = len(st_207)
        
# restrict spatio-temporal sequences to length-5
STC5_207 = [ st_207 for st_207 in STC_207 if len(st_207) <= 5 ]

# unpack spatio-temporal sequences to temporal sequences
TC5_207 = []
for n in range(len(STC5_207)):
    tc_207 = [ u for u in STC5_207[n][0] ]
    for k in range(1, len(STC5_207[n])):
        tc_207 = [ u+v for u in tc_207 for v in STC5_207[n][k] ]
    TC5_207 = TC5_207 + tc_207
setTC5_207 = list(set(TC5_207))

# get empirical frequencies
fTC5_207 = [ TC5_207.count(setTC5_207[l]) for l in range(len(setTC5_207)) ]

# statistical testing against null model
np.random.seed(0)
fTC5null_207 = [ [] for l in range(len(setTC5_207)) ]
for s in range(207):
    STC5null_207 = [ list(st_207) for st_207 in STC5_207]
    for st_207 in STC5null_207:
        np.random.shuffle(st_207)
    TC5null_207= []
    for n in range(len(STC5null_207)):
        tc_207 = [ u for u in STC5null_207[n][0] ]
        for k in range(1, len(STC5null_207[n])):
            tc_207 = [ u+v for u in tc_207 for v in STC5null_207[n][k] ]
        TC5null_207= TC5null_207+ tc_207
    for l in range(len(setTC5_207)):
        if TC5null_207.count(setTC5_207[l]) > 0:
            fTC5null_207[l].append(TC5null_207.count(setTC5_207[l]))
        else:
            fTC5null_207[l].append(0)
```

```python
print("List of Statistically Significant Sequences at α = 0.05")

#statistical significance testing at p < 0.05 level
for l in range(len(setTC5_207)):
    if sum(freq >= fTC5_207[l] for freq in fTC5null_207[l]) < 5:
        print(setTC5_207[l], fTC5_207[l], sum(freq >= fTC5_207[l] for freq in fTC5null_207[l]))
```

```python
print("List of Statistically Significant Sequences at α = 0.01")

#statistical significance testing at p < 0.01 level
for l in range(len(setTC5_207)):
    if sum(freq >= fTC5_207[l] for freq in fTC5null_207[l]) < 1:
        print(setTC5_207[l], fTC5_207[l], sum(freq >= fTC5_207[l] for freq in fTC5null_207[l]))
```
