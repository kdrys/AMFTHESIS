# 76 (L) GTCAP - (β) SM - (E) ALI - (T) MBT - (V) TEL

```python
#extract the overlapping dates

start_76 = max(min(Date_GTCAP), min(Date_SM), min(Date_ALI), min(Date_MBT), min(Date_TEL))
end_76 = min(max(Date_GTCAP), max(Date_SM), max(Date_ALI), max(Date_MBT), max(Date_TEL))

#shorten the closing prices
GTCAP_76_1 = Date_GTCAP.index(start_76)
GTCAP_76_2 = Date_GTCAP.index(end_76)

SM_76_1 = Date_SM.index(start_76)
SM_76_2 = Date_SM.index(end_76)

ALI_76_1 = Date_ALI.index(start_76)
ALI_76_2 = Date_ALI.index(end_76)

MBT_76_1 = Date_MBT.index(start_76)
MBT_76_2 = Date_MBT.index(end_76)

TEL_76_1 = Date_TEL.index(start_76)
TEL_76_2 = Date_TEL.index(end_76)

#save the shortened versions of closing prices
L76 = np.save('GTCAPshortPrices.npy', ClosePrice_GTCAP[GTCAP_76_1:GTCAP_76_2])
β76 = np.save('SMshortPrices.npy', ClosePrice_SM[SM_76_1:SM_76_2])
E76 = np.save('ALIshortPrices.npy', ClosePrice_ALI[ALI_76_1:ALI_76_2])
T76 = np.save('MBTshortPrices.npy', ClosePrice_MBT[MBT_76_1:MBT_76_2])
V76 = np.save('TELshortPrices.npy', ClosePrice_TEL[TEL_76_1:TEL_76_2])

# import modules
import numpy as np
# import prices
L76 = np.load('GTCAPshortPrices.npy', allow_pickle=True).tolist()
β76 = np.load('SMshortPrices.npy', allow_pickle=True).tolist()
E76 = np.load('ALIshortPrices.npy', allow_pickle=True).tolist()
T76 = np.load('MBTshortPrices.npy', allow_pickle=True).tolist()
V76 = np.load('TELshortPrices.npy', allow_pickle=True).tolist()
```

```python
# extract spatio-temporal sequences
STC_76 = []
st_76 = []
n = 1
while n < len(L76):
    sc_76 = ''
    if L76[n] > L76[n-1]:
        sc_76 = sc_76 + 'L'
    if β76[n] > β76[n-1]:
        sc_76 = sc_76 + 'β'
    if E76[n] > E76[n-1]:
        sc_76 = sc_76 + 'E'
    if T76[n] > T76[n-1]:
        sc_76 = sc_76 + 'T'
    if V76[n] > V76[n-1]:
        sc_76 = sc_76 + 'V'
    if len(sc_76) > 0:
        st_76.append(sc_76)
    else:
        STC_76.append(st_76)
        st_76 = []
    n = n + 1


# record the number of empty spatio-temporal sequences
nSTCempty_76 = STC_76.count([])
# keep only non-empty spatio-temporal sequences
STC_76 = [ st_76 for st_76 in STC_76 if len(st_76) > 0 ]

# record the length of the longest spatio-temporal sequence
lmin_76 = len(L76)
lmax_76 = 0
for st_76 in STC_76:
    if len(st_76) < lmin_76:
        lmin_76 = len(st_76)
    if len(st_76) > lmax_76:
        lmax_76 = len(st_76)
        
# restrict spatio-temporal sequences to length-5
STC5_76 = [ st_76 for st_76 in STC_76 if len(st_76) <= 5 ]

# unpack spatio-temporal sequences to temporal sequences
TC5_76 = []
for n in range(len(STC5_76)):
    tc_76 = [ u for u in STC5_76[n][0] ]
    for k in range(1, len(STC5_76[n])):
        tc_76 = [ u+v for u in tc_76 for v in STC5_76[n][k] ]
    TC5_76 = TC5_76 + tc_76
setTC5_76 = list(set(TC5_76))

# get empirical frequencies
fTC5_76 = [ TC5_76.count(setTC5_76[l]) for l in range(len(setTC5_76)) ]

# statistical testing against null model
np.random.seed(0)
fTC5null_76 = [ [] for l in range(len(setTC5_76)) ]
for s in range(100):
    STC5null_76 = [ list(st_76) for st_76 in STC5_76]
    for st_76 in STC5null_76:
        np.random.shuffle(st_76)
    TC5null_76= []
    for n in range(len(STC5null_76)):
        tc_76 = [ u for u in STC5null_76[n][0] ]
        for k in range(1, len(STC5null_76[n])):
            tc_76 = [ u+v for u in tc_76 for v in STC5null_76[n][k] ]
        TC5null_76= TC5null_76+ tc_76
    for l in range(len(setTC5_76)):
        if TC5null_76.count(setTC5_76[l]) > 0:
            fTC5null_76[l].append(TC5null_76.count(setTC5_76[l]))
        else:
            fTC5null_76[l].append(0)
```

```python
print("List of Statistically Significant Sequences at α = 0.05")

#statistical significance testing at p < 0.05 level
for l in range(len(setTC5_76)):
    if sum(freq >= fTC5_76[l] for freq in fTC5null_76[l]) < 5:
        print(setTC5_76[l])
```

```python
print("List of Statistically Significant Sequences at α = 0.01")

#statistical significance testing at p < 0.01 level
for l in range(len(setTC5_76)):
    if sum(freq >= fTC5_76[l] for freq in fTC5null_76[l]) < 1:
        print(setTC5_76[l]) 
```

```python
# extract spatio-temporal sequences
STC_87 = []
st_87 = []
n = 1
while n < len(L87):
    sc_87 = ''
    if L87[n] > L87[n-1]:
        sc_87 = sc_87 + 'L'
    if β87[n] > β87[n-1]:
        sc_87 = sc_87 + 'β'
    if Z87[n] > Z87[n-1]:
        sc_87 = sc_87 + 'Z'
    if ε87[n] > ε87[n-1]:
        sc_87 = sc_87 + 'ε'
    if W87[n] > W87[n-1]:
        sc_87 = sc_87 + 'W'
    if len(sc_87) > 0:
        st_87.append(sc_87)
    else:
        STC_87.append(st_87)
        st_87 = []
    n = n + 1


# record the number of empty spatio-temporal sequences
nSTCempty_87 = STC_87.count([])
# keep only non-empty spatio-temporal sequences
STC_87 = [ st_87 for st_87 in STC_87 if len(st_87) > 0 ]

# record the length of the longest spatio-temporal sequence
lmin_87 = len(L87)
lmax_87 = 0
for st_87 in STC_87:
    if len(st_87) < lmin_87:
        lmin_87 = len(st_87)
    if len(st_87) > lmax_87:
        lmax_87 = len(st_87)
        
# restrict spatio-temporal sequences to length-5
STC5_87 = [ st_87 for st_87 in STC_87 if len(st_87) <= 5 ]

# unpack spatio-temporal sequences to temporal sequences
TC5_87 = []
for n in range(len(STC5_87)):
    tc_87 = [ u for u in STC5_87[n][0] ]
    for k in range(1, len(STC5_87[n])):
        tc_87 = [ u+v for u in tc_87 for v in STC5_87[n][k] ]
    TC5_87 = TC5_87 + tc_87
setTC5_87 = list(set(TC5_87))

# get empirical frequencies
fTC5_87 = [ TC5_87.count(setTC5_87[l]) for l in range(len(setTC5_87)) ]

# statistical testing against null model
np.random.seed(0)
fTC5null_87 = [ [] for l in range(len(setTC5_87)) ]
for s in range(100):
    STC5null_87 = [ list(st_87) for st_87 in STC5_87]
    for st_87 in STC5null_87:
        np.random.shuffle(st_87)
    TC5null_87= []
    for n in range(len(STC5null_87)):
        tc_87 = [ u for u in STC5null_87[n][0] ]
        for k in range(1, len(STC5null_87[n])):
            tc_87 = [ u+v for u in tc_87 for v in STC5null_87[n][k] ]
        TC5null_87= TC5null_87+ tc_87
    for l in range(len(setTC5_87)):
        if TC5null_87.count(setTC5_87[l]) > 0:
            fTC5null_87[l].append(TC5null_87.count(setTC5_87[l]))
        else:
            fTC5null_87[l].append(0)
```

```python
print("List of Statistically Significant Sequences at α = 0.05")

#statistical significance testing at p < 0.05 level
for l in range(len(setTC5_87)):
    if sum(freq >= fTC5_87[l] for freq in fTC5null_87[l]) < 5:
        print(setTC5_87[l])
```

```python
print("List of Statistically Significant Sequences at α = 0.01")

#statistical significance testing at p < 0.01 level
for l in range(len(setTC5_87)):
    if sum(freq >= fTC5_87[l] for freq in fTC5null_87[l]) < 1:
        print(setTC5_87[l])
```
