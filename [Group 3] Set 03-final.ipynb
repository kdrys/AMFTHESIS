```python
#shorten the closing prices
AP_31_1 = Date_AP.index(start_31)
AP_31_2 = Date_AP.index(end_31)

AC_31_1 = Date_AC.index(start_31)
AC_31_2 = Date_AC.index(end_31)

BLOOM_31_1 = Date_BLOOM.index(start_31)
BLOOM_31_2 = Date_BLOOM.index(end_31)

JGS_31_1 = Date_JGS.index(start_31)
JGS_31_2 = Date_JGS.index(end_31)

MEG_31_1 = Date_MEG.index(start_31)
MEG_31_2 = Date_MEG.index(end_31)
```

```python
# import modules
import numpy as np
# import prices

B31 = np.load('APshortPrices.npy', allow_pickle=True).tolist() 
D31 = np.load('ACshortPrices.npy', allow_pickle=True).tolist() 
H31 = np.load('BLOOMshortPrices.npy', allow_pickle=True).tolist() 
N31 = np.load('JGSshortPrices.npy', allow_pickle=True).tolist() 
R31 = np.load('MEGshortPrices.npy', allow_pickle=True).tolist()
```

```python
# extract spatio-temporal sequences
STC_31 = []
st_31 = []
n = 1
while n < len(B31):
    sc_31 = ''
    if B31[n] > B31[n-1]:
        sc_31 = sc_31 + 'B'
    if D31[n] > D31[n-1]:
        sc_31 = sc_31 + 'D'
    if H31[n] > H31[n-1]:
        sc_31 = sc_31 + 'H'
    if N31[n] > N31[n-1]:
        sc_31 = sc_31 + 'N'
    if R31[n] > R31[n-1]:
        sc_31 = sc_31 + 'R'
    if len(sc_31) > 0:
        st_31.append(sc_31)
    else:
        STC_31.append(st_31)
        st_31 = []
    n = n + 1
```

```python
# record the number of empty spatio-temporal sequences
nSTCempty_31 = STC_31.count([])
# keep only non-empty spatio-temporal sequences
STC_31 = [ st_31 for st_31 in STC_31 if len(st_31) > 0 ]

# record the length of the longest spatio-temporal sequence
lmin_31 = len(B31)
lmax_31 = 0
for st_31 in STC_31:
    if len(st_31) < lmin_31:
        lmin_31 = len(st_31)
    if len(st_31) > lmax_31:
        lmax_31 = len(st_31)
        
# restrict spatio-temporal sequences to length-5
STC5_31 = [ st_31 for st_31 in STC_31 if len(st_31) <= 5 ]

# unpack spatio-temporal sequences to temporal sequences
TC5_31 = []
for n in range(len(STC5_31)):
    tc_31 = [ u for u in STC5_31[n][0] ]
    for k in range(1, len(STC5_31[n])):
        tc_31 = [ u+v for u in tc_31 for v in STC5_31[n][k] ]
    TC5_31 = TC5_31 + tc_31
setTC5_31 = list(set(TC5_31))

# get empirical frequencies
fTC5_31 = [ TC5_31.count(setTC5_31[l]) for l in range(len(setTC5_31)) ]

# statistical testing against null model
np.random.seed(0)
fTC5null_31 = [ [] for l in range(len(setTC5_31)) ]
for s in range(100):
    STC5null_31 = [ list(st_31) for st_31 in STC5_31]
    for st_31 in STC5null_31:
        np.random.shuffle(st_31)
    TC5null_31= []
    for n in range(len(STC5null_31)):
        tc_31 = [ u for u in STC5null_31[n][0] ]
        for k in range(1, len(STC5null_31[n])):
            tc_31 = [ u+v for u in tc_31 for v in STC5null_31[n][k] ]
        TC5null_31= TC5null_31+ tc_31
    for l in range(len(setTC5_31)):
        if TC5null_31.count(setTC5_31[l]) > 0:
            fTC5null_31[l].append(TC5null_31.count(setTC5_31[l]))
        else:
            fTC5null_31[l].append(0)
```

```python
print("List of Statistically Significant Sequences at R = 0.05")

#statistical significance testing at p < 0.05 level
for l in range(len(setTC5_31)):
     if sum(freq >= fTC5_31[l] for freq in fTC5null_31[l]) < 5:
        print(setTC5_31[l], fTC5_31[l], sum(freq >= fTC5_31[l] for freq in fTC5null_31[l]))
#         print(setTC5_31[l])
```

```python
print("List of Statistically Significant Sequences at R = 0.01")

#statistical significance testing at p < 0.01 level
for l in range(len(setTC5_31)):
     if sum(freq >= fTC5_31[l] for freq in fTC5null_31[l]) < 1:
        print(setTC5_31[l], fTC5_31[l], sum(freq >= fTC5_31[l] for freq in fTC5null_31[l]))
#         print(setTC5_31[l])
```
