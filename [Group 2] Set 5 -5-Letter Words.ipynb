# Five-letter Words

# 61 (L) GTCAP - (β) SM - (Q) MER - (E) ALI - (W) PGOLD

```python
#extract the overlapping dates

start_61 = max(min(Date_GTCAP), min(Date_SM), min(Date_MER), min(Date_ALI), min(Date_PGOLD))
end_61 = min(max(Date_GTCAP), max(Date_SM), max(Date_MER), max(Date_ALI), max(Date_PGOLD))

#shorten the closing prices
GTCAP_61_1 = Date_GTCAP.index(start_61)
GTCAP_61_2 = Date_GTCAP.index(end_61)

SM_61_1 = Date_SM.index(start_61)
SM_61_2 = Date_SM.index(end_61)

MER_61_1 = Date_MER.index(start_61)
MER_61_2 = Date_MER.index(end_61)

ALI_61_1 = Date_ALI.index(start_61)
ALI_61_2 = Date_ALI.index(end_61)

PGOLD_61_1 = Date_PGOLD.index(start_61)
PGOLD_61_2 = Date_PGOLD.index(end_61)

#save the shortened versions of closing prices
L61 = np.save('GTCAPshortPrices.npy', ClosePrice_GTCAP[GTCAP_61_1:GTCAP_61_2])
β61 = np.save('SMshortPrices.npy', ClosePrice_SM[SM_61_1:SM_61_2])
Q61 = np.save('MERshortPrices.npy', ClosePrice_MER[MER_61_1:MER_61_2])
E61 = np.save('ALIshortPrices.npy', ClosePrice_ALI[ALI_61_1:ALI_61_2])
W61 = np.save('PGOLDshortPrices.npy', ClosePrice_PGOLD[PGOLD_61_1:PGOLD_61_2])

# import modules
import numpy as np
# import prices
L61 = np.load('GTCAPshortPrices.npy', allow_pickle=True).tolist()
β61 = np.load('SMshortPrices.npy', allow_pickle=True).tolist()
Q61 = np.load('MERshortPrices.npy', allow_pickle=True).tolist()
E61 = np.load('ALIshortPrices.npy', allow_pickle=True).tolist()
W61 = np.load('PGOLDshortPrices.npy', allow_pickle=True).tolist()
```

```python
# extract spatio-temporal sequences
STC_61 = []
st_61 = []
n = 1
while n < len(L61):
    sc_61 = ''
    if L61[n] > L61[n-1]:
        sc_61 = sc_61 + 'L'
    if β61[n] > β61[n-1]:
        sc_61 = sc_61 + 'β'
    if Q61[n] > Q61[n-1]:
        sc_61 = sc_61 + 'Q'
    if E61[n] > E61[n-1]:
        sc_61 = sc_61 + 'E'
    if W61[n] > W61[n-1]:
        sc_61 = sc_61 + 'W'
    if len(sc_61) > 0:
        st_61.append(sc_61)
    else:
        STC_61.append(st_61)
        st_61 = []
    n = n + 1


# record the number of empty spatio-temporal sequences
nSTCempty_61 = STC_61.count([])
# keep only non-empty spatio-temporal sequences
STC_61 = [ st_61 for st_61 in STC_61 if len(st_61) > 0 ]

# record the length of the longest spatio-temporal sequence
lmin_61 = len(L61)
lmax_61 = 0
for st_61 in STC_61:
    if len(st_61) < lmin_61:
        lmin_61 = len(st_61)
    if len(st_61) > lmax_61:
        lmax_61 = len(st_61)
        
# restrict spatio-temporal sequences to length-5
STC5_61 = [ st_61 for st_61 in STC_61 if len(st_61) <= 5 ]

# unpack spatio-temporal sequences to temporal sequences
TC5_61 = []
for n in range(len(STC5_61)):
    tc_61 = [ u for u in STC5_61[n][0] ]
    for k in range(1, len(STC5_61[n])):
        tc_61 = [ u+v for u in tc_61 for v in STC5_61[n][k] ]
    TC5_61 = TC5_61 + tc_61
setTC5_61 = list(set(TC5_61))

# get empirical frequencies
fTC5_61 = [ TC5_61.count(setTC5_61[l]) for l in range(len(setTC5_61)) ]

# statistical testing against null model
np.random.seed(0)
fTC5null_61 = [ [] for l in range(len(setTC5_61)) ]
for s in range(100):
    STC5null_61 = [ list(st_61) for st_61 in STC5_61]
    for st_61 in STC5null_61:
        np.random.shuffle(st_61)
    TC5null_61= []
    for n in range(len(STC5null_61)):
        tc_61 = [ u for u in STC5null_61[n][0] ]
        for k in range(1, len(STC5null_61[n])):
            tc_61 = [ u+v for u in tc_61 for v in STC5null_61[n][k] ]
        TC5null_61= TC5null_61+ tc_61
    for l in range(len(setTC5_61)):
        if TC5null_61.count(setTC5_61[l]) > 0:
            fTC5null_61[l].append(TC5null_61.count(setTC5_61[l]))
        else:
            fTC5null_61[l].append(0)
```

```python
print("List of Statistically Significant Sequences at α = 0.05")

#statistical significance testing at p < 0.05 level
for l in range(len(setTC5_61)):
    if sum(freq >= fTC5_61[l] for freq in fTC5null_61[l]) < 5:
        print(setTC5_61[l], fTC5_61[l], sum(freq >= fTC5_61[l] for freq in fTC5null_61[l]))
```

```python
print("List of Statistically Significant Sequences at α = 0.01")

#statistical significance testing at p < 0.01 level
for l in range(len(setTC5_61)):
    if sum(freq >= fTC5_61[l] for freq in fTC5null_61[l]) < 1:
        print(setTC5_61[l], fTC5_61[l], sum(freq >= fTC5_61[l] for freq in fTC5null_61[l]))
```

```python
# extract spatio-temporal sequences
STC_72 = []
st_72 = []
n = 1
while n < len(L72):
    sc_72 = ''
    if L72[n] > L72[n-1]:
        sc_72 = sc_72 + 'L'
    if β72[n] > β72[n-1]:
        sc_72 = sc_72 + 'β'
    if E72[n] > E72[n-1]:
        sc_72 = sc_72 + 'E'
    if Z72[n] > Z72[n-1]:
        sc_72 = sc_72 + 'Z'
    if T72[n] > T72[n-1]:
        sc_72 = sc_72 + 'T'
    if len(sc_72) > 0:
        st_72.append(sc_72)
    else:
        STC_72.append(st_72)
        st_72 = []
    n = n + 1


# record the number of empty spatio-temporal sequences
nSTCempty_72 = STC_72.count([])
# keep only non-empty spatio-temporal sequences
STC_72 = [ st_72 for st_72 in STC_72 if len(st_72) > 0 ]

# record the length of the longest spatio-temporal sequence
lmin_72 = len(L72)
lmax_72 = 0
for st_72 in STC_72:
    if len(st_72) < lmin_72:
        lmin_72 = len(st_72)
    if len(st_72) > lmax_72:
        lmax_72 = len(st_72)
        
# restrict spatio-temporal sequences to length-5
STC5_72 = [ st_72 for st_72 in STC_72 if len(st_72) <= 5 ]

# unpack spatio-temporal sequences to temporal sequences
TC5_72 = []
for n in range(len(STC5_72)):
    tc_72 = [ u for u in STC5_72[n][0] ]
    for k in range(1, len(STC5_72[n])):
        tc_72 = [ u+v for u in tc_72 for v in STC5_72[n][k] ]
    TC5_72 = TC5_72 + tc_72
setTC5_72 = list(set(TC5_72))

# get empirical frequencies
fTC5_72 = [ TC5_72.count(setTC5_72[l]) for l in range(len(setTC5_72)) ]

# statistical testing against null model
np.random.seed(0)
fTC5null_72 = [ [] for l in range(len(setTC5_72)) ]
for s in range(100):
    STC5null_72 = [ list(st_72) for st_72 in STC5_72]
    for st_72 in STC5null_72:
        np.random.shuffle(st_72)
    TC5null_72= []
    for n in range(len(STC5null_72)):
        tc_72 = [ u for u in STC5null_72[n][0] ]
        for k in range(1, len(STC5null_72[n])):
            tc_72 = [ u+v for u in tc_72 for v in STC5null_72[n][k] ]
        TC5null_72= TC5null_72+ tc_72
    for l in range(len(setTC5_72)):
        if TC5null_72.count(setTC5_72[l]) > 0:
            fTC5null_72[l].append(TC5null_72.count(setTC5_72[l]))
        else:
            fTC5null_72[l].append(0)
```

```python
print("List of Statistically Significant Sequences at α = 0.05")

#statistical significance testing at p < 0.05 level
for l in range(len(setTC5_72)):
    if sum(freq >= fTC5_72[l] for freq in fTC5null_72[l]) < 5:
        print(setTC5_72[l], fTC5_72[l], sum(freq >= fTC5_72[l] for freq in fTC5null_72[l]))
```

```python
print("List of Statistically Significant Sequences at α = 0.01")

#statistical significance testing at p < 0.01 level
for l in range(len(setTC5_72)):
    if sum(freq >= fTC5_72[l] for freq in fTC5null_72[l]) < 1:
        print(setTC5_72[l], fTC5_72[l], sum(freq >= fTC5_72[l] for freq in fTC5null_72[l]))
```
