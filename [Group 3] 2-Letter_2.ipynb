```python
#which dataset is longer

if len(D) > len(Y):
    print("AC dataset is longer.")
elif len(Y) > len(D):
    print("SMC dataset is longer.")
else:
    print("The datasets have the same number of data points.")
```

```python
N1Y = DateY.index(max(min(DateY),min(DateD)))
N2Y = DateY.index(min(max(DateY),max(DateD)))
Y1 = Y[N1Y:N2Y +1]

STC_DY  = [] 
st_DY  = [] 
n = 1
while n < len(Y1):
    sc = ''
    if Y1[n] > Y1[n-1]:  
        sc = sc + 'Y' 
    if D[n] > D[n-1]: 
        sc = sc + 'D'
    if len(sc) > 0:
        st_DY.append(sc) 
    else:
        STC_DY.append(st_DY ) 
        st_DY  = []
    n = n + 1

nSTCempty_DY  = STC_DY .count([])

STC_DY  = [ st_DY  for st_DY  in STC_DY  if len(st_DY ) > 0 ]

lmax_DY  = 0
for st_DY  in STC_DY :
    if len(st_DY ) > lmax_DY :
        lmax_DY  = len(st_DY )

STC5_DY  = [ st_DY   for st_DY   in STC_DY   if len(st_DY ) <= 5 ]

TC5_DY  = [] 
for n in range(len(STC5_DY )):
    tc_DY   = [ u for u in STC5_DY  [n][0] ]
    for k in range(1, len(STC5_DY [n])):
        tc_DY   = [ u+v for u in tc_DY   for v in STC5_DY  [n][k] ]
    TC5_DY  = TC5_DY  + tc_DY  
setTC5_DY   = list(set(TC5_DY ))


fTC5_DY  = [ TC5_DY .count(setTC5_DY [l]) for l in range(len(setTC5_DY )) ]

fmaxTC5_DY  = [ 0 for l in range(len(setTC5_DY )) ]
for s in range(100):
    STC5null_DY  = [ list(st_DY ) for st_DY  in STC5_DY  ]
    for st_DY  in STC5null_DY :
        np.random.shuffle(st_DY )
    TC5null_DY  = []
    for n in range(len(STC5null_DY )):
        tc = [ u for u in STC5null_DY [n][0] ]
        for k in range(1, len(STC5null_DY [n])):
            tc = [ u+v for u in tc_DY  for v in STC5null_DY [n][k] ]
        TC5null_DY  = TC5null_DY  + tc_DY 
        
for l in range(len(setTC5_DY )):
    if TC5null_DY .count(setTC5_DY [l]) > fTC5_DY [l]:
        fmaxTC5_DY [l] += 1
```

```python
N1R = DateR.index(max(min(DateR),min(DateH)))
N2R = DateR.index(min(max(DateR),max(DateH)))
R1 = R[N1R:N2R +1]

STC_HR  = [] 
st_HR  = [] 
n = 1
while n < len(R1):
    sc = ''
    if R1[n] > R1[n-1]:  
        sc = sc + 'R' 
    if H[n] > H[n-1]: 
        sc = sc + 'H'
    if len(sc) > 0:
        st_HR.append(sc) 
    else:
        STC_HR.append(st_HR ) 
        st_HR  = []
    n = n + 1

nSTCempty_HR  = STC_HR .count([])

STC_HR  = [ st_HR  for st_HR  in STC_HR  if len(st_HR ) > 0 ]

lmax_HR  = 0
for st_HR  in STC_HR :
    if len(st_HR ) > lmax_HR :
        lmax_HR  = len(st_HR )

STC5_HR  = [ st_HR   for st_HR   in STC_HR   if len(st_HR ) <= 5 ]

TC5_HR  = [] 
for n in range(len(STC5_HR )):
    tc_HR   = [ u for u in STC5_HR  [n][0] ]
    for k in range(1, len(STC5_HR [n])):
        tc_HR   = [ u+v for u in tc_HR   for v in STC5_HR  [n][k] ]
    TC5_HR  = TC5_HR  + tc_HR  
setTC5_HR   = list(set(TC5_HR ))


fTC5_HR  = [ TC5_HR .count(setTC5_HR [l]) for l in range(len(setTC5_HR )) ]

fmaxTC5_HR  = [ 0 for l in range(len(setTC5_HR )) ]
for s in range(100):
    STC5null_HR  = [ list(st_HR ) for st_HR  in STC5_HR  ]
    for st_HR  in STC5null_HR :
        np.random.shuffle(st_HR )
    TC5null_HR  = []
    for n in range(len(STC5null_HR )):
        tc = [ u for u in STC5null_HR [n][0] ]
        for k in range(1, len(STC5null_HR [n])):
            tc = [ u+v for u in tc_HR  for v in STC5null_HR [n][k] ]
        TC5null_HR  = TC5null_HR  + tc_HR 
        
for l in range(len(setTC5_HR )):
    if TC5null_HR .count(setTC5_HR [l]) > fTC5_HR [l]:
        fmaxTC5_HR [l] += 1
```

```python
print("List of Statistically Significant Sequences at α = 0.05: ")

for l in range(len(setTC5_HR )):
    if fmaxTC5_HR [l] < 5: #0.05
        print(setTC5_HR [l])
```

```python
print("List of Statistically Significant Sequences at α = 0.01: ")
        
for l in range(len(setTC5_HR )):
    if fmaxTC5_HR [l] < 1: #0.01
        print(setTC5_HR [l])  
```
