# 136 (C) AGI - (β) SM - (Q) MER - (T) MBT - (V) TEL

```python
#extract the overlapping dates

start_136 = max(min(Date_AGI), min(Date_SM), min(Date_MER), min(Date_MBT), min(Date_TEL))
end_136 = min(max(Date_AGI), max(Date_SM), max(Date_MER), max(Date_MBT), max(Date_TEL))

#shorten the closing prices
AGI_136_1 = Date_AGI.index(start_136)
AGI_136_2 = Date_AGI.index(end_136)

SM_136_1 = Date_SM.index(start_136)
SM_136_2 = Date_SM.index(end_136)

MER_136_1 = Date_MER.index(start_136)
MER_136_2 = Date_MER.index(end_136)

MBT_136_1 = Date_MBT.index(start_136)
MBT_136_2 = Date_MBT.index(end_136)

TEL_136_1 = Date_TEL.index(start_136)
TEL_136_2 = Date_TEL.index(end_136)

#save the shortened versions of closing prices
C136 = np.save('AGIshortPrices.npy', ClosePrice_AGI[AGI_136_1:AGI_136_2])
β136 = np.save('SMshortPrices.npy', ClosePrice_SM[SM_136_1:SM_136_2])
Q136 = np.save('MERshortPrices.npy', ClosePrice_MER[MER_136_1:MER_136_2])
T136 = np.save('MBTshortPrices.npy', ClosePrice_MBT[MBT_136_1:MBT_136_2])
V136 = np.save('TELshortPrices.npy', ClosePrice_TEL[TEL_136_1:TEL_136_2])

# import modules
import numpy as np
# import prices
C136 = np.load('AGIshortPrices.npy', allow_pickle=True).tolist()
β136 = np.load('SMshortPrices.npy', allow_pickle=True).tolist()
Q136 = np.load('MERshortPrices.npy', allow_pickle=True).tolist()
T136 = np.load('MBTshortPrices.npy', allow_pickle=True).tolist()
V136 = np.load('TELshortPrices.npy', allow_pickle=True).tolist()
```

```python
# extract spatio-temporal sequences
STC_136 = []
st_136 = []
n = 1
while n < len(C136):
    sc_136 = ''
    if C136[n] > C136[n-1]:
        sc_136 = sc_136 + 'C'
    if β136[n] > β136[n-1]:
        sc_136 = sc_136 + 'β'
    if Q136[n] > Q136[n-1]:
        sc_136 = sc_136 + 'Q'
    if T136[n] > T136[n-1]:
        sc_136 = sc_136 + 'T'
    if V136[n] > V136[n-1]:
        sc_136 = sc_136 + 'V'
    if len(sc_136) > 0:
        st_136.append(sc_136)
    else:
        STC_136.append(st_136)
        st_136 = []
    n = n + 1


# record the number of empty spatio-temporal sequences
nSTCempty_136 = STC_136.count([])
# keep only non-empty spatio-temporal sequences
STC_136 = [ st_136 for st_136 in STC_136 if len(st_136) > 0 ]

# record the length of the longest spatio-temporal sequence
lmin_136 = len(C136)
lmax_136 = 0
for st_136 in STC_136:
    if len(st_136) < lmin_136:
        lmin_136 = len(st_136)
    if len(st_136) > lmax_136:
        lmax_136 = len(st_136)
        
# restrict spatio-temporal sequences to length-5
STC5_136 = [ st_136 for st_136 in STC_136 if len(st_136) <= 5 ]

# unpack spatio-temporal sequences to temporal sequences
TC5_136 = []
for n in range(len(STC5_136)):
    tc_136 = [ u for u in STC5_136[n][0] ]
    for k in range(1, len(STC5_136[n])):
        tc_136 = [ u+v for u in tc_136 for v in STC5_136[n][k] ]
    TC5_136 = TC5_136 + tc_136
setTC5_136 = list(set(TC5_136))

# get empirical frequencies
fTC5_136 = [ TC5_136.count(setTC5_136[l]) for l in range(len(setTC5_136)) ]

# statistical testing against null model
np.random.seed(0)
fTC5null_136 = [ [] for l in range(len(setTC5_136)) ]
for s in range(136):
    STC5null_136 = [ list(st_136) for st_136 in STC5_136]
    for st_136 in STC5null_136:
        np.random.shuffle(st_136)
    TC5null_136= []
    for n in range(len(STC5null_136)):
        tc_136 = [ u for u in STC5null_136[n][0] ]
        for k in range(1, len(STC5null_136[n])):
            tc_136 = [ u+v for u in tc_136 for v in STC5null_136[n][k] ]
        TC5null_136= TC5null_136+ tc_136
    for l in range(len(setTC5_136)):
        if TC5null_136.count(setTC5_136[l]) > 0:
            fTC5null_136[l].append(TC5null_136.count(setTC5_136[l]))
        else:
            fTC5null_136[l].append(0)
```

```python
print("List of Statistically Significant Sequences at α = 0.05")

#statistical significance testing at p < 0.05 level
for l in range(len(setTC5_136)):
    if sum(freq >= fTC5_136[l] for freq in fTC5null_136[l]) < 5:
        print(setTC5_136[l], fTC5_136[l], sum(freq >= fTC5_136[l] for freq in fTC5null_136[l]))
```

```python
print("List of Statistically Significant Sequences at α = 0.01")

#statistical significance testing at p < 0.01 level
for l in range(len(setTC5_136)):
    if sum(freq >= fTC5_136[l] for freq in fTC5null_136[l]) < 1:
        print(setTC5_136[l], fTC5_136[l], sum(freq >= fTC5_136[l] for freq in fTC5null_136[l]))
```

```python
# extract spatio-temporal sequences
STC_147 = []
st_147 = []
n = 1
while n < len(C147):
    sc_147 = ''
    if C147[n] > C147[n-1]:
        sc_147 = sc_147 + 'C'
    if β147[n] > β147[n-1]:
        sc_147 = sc_147 + 'β'
    if E147[n] > E147[n-1]:
        sc_147 = sc_147 + 'E'
    if T147[n] > T147[n-1]:
        sc_147 = sc_147 + 'T'
    if ε147[n] > ε147[n-1]:
        sc_147 = sc_147 + 'ε'
    if len(sc_147) > 0:
        st_147.append(sc_147)
    else:
        STC_147.append(st_147)
        st_147 = []
    n = n + 1


# record the number of empty spatio-temporal sequences
nSTCempty_147 = STC_147.count([])
# keep only non-empty spatio-temporal sequences
STC_147 = [ st_147 for st_147 in STC_147 if len(st_147) > 0 ]

# record the length of the longest spatio-temporal sequence
lmin_147 = len(C147)
lmax_147 = 0
for st_147 in STC_147:
    if len(st_147) < lmin_147:
        lmin_147 = len(st_147)
    if len(st_147) > lmax_147:
        lmax_147 = len(st_147)
        
# restrict spatio-temporal sequences to length-5
STC5_147 = [ st_147 for st_147 in STC_147 if len(st_147) <= 5 ]

# unpack spatio-temporal sequences to temporal sequences
TC5_147 = []
for n in range(len(STC5_147)):
    tc_147 = [ u for u in STC5_147[n][0] ]
    for k in range(1, len(STC5_147[n])):
        tc_147 = [ u+v for u in tc_147 for v in STC5_147[n][k] ]
    TC5_147 = TC5_147 + tc_147
setTC5_147 = list(set(TC5_147))

# get empirical frequencies
fTC5_147 = [ TC5_147.count(setTC5_147[l]) for l in range(len(setTC5_147)) ]

# statistical testing against null model
np.random.seed(0)
fTC5null_147 = [ [] for l in range(len(setTC5_147)) ]
for s in range(147):
    STC5null_147 = [ list(st_147) for st_147 in STC5_147]
    for st_147 in STC5null_147:
        np.random.shuffle(st_147)
    TC5null_147= []
    for n in range(len(STC5null_147)):
        tc_147 = [ u for u in STC5null_147[n][0] ]
        for k in range(1, len(STC5null_147[n])):
            tc_147 = [ u+v for u in tc_147 for v in STC5null_147[n][k] ]
        TC5null_147= TC5null_147+ tc_147
    for l in range(len(setTC5_147)):
        if TC5null_147.count(setTC5_147[l]) > 0:
            fTC5null_147[l].append(TC5null_147.count(setTC5_147[l]))
        else:
            fTC5null_147[l].append(0)
```

```python
print("List of Statistically Significant Sequences at α = 0.05")

#statistical significance testing at p < 0.05 level
for l in range(len(setTC5_147)):
    if sum(freq >= fTC5_147[l] for freq in fTC5null_147[l]) < 5:
        print(setTC5_147[l], fTC5_147[l], sum(freq >= fTC5_147[l] for freq in fTC5null_147[l]))
```

```python
print("List of Statistically Significant Sequences at α = 0.01")

#statistical significance testing at p < 0.01 level
for l in range(len(setTC5_147)):
    if sum(freq >= fTC5_147[l] for freq in fTC5null_147[l]) < 1:
        print(setTC5_147[l], fTC5_147[l], sum(freq >= fTC5_147[l] for freq in fTC5null_147[l]))
```
