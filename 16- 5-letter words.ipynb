# 226 (β) SM - (E) ALI - (V) TEL - (ε) URC - (W) PGOLD

```python
#extract the overlapping dates

start_226 = max(min(Date_SM), min(Date_ALI), min(Date_TEL), min(Date_URC), min(Date_PGOLD))
end_226 = min(max(Date_SM), max(Date_ALI), max(Date_TEL), max(Date_URC), max(Date_PGOLD))

#shorten the closing prices
SM_226_1 = Date_SM.index(start_226)
SM_226_2 = Date_SM.index(end_226)

ALI_226_1 = Date_ALI.index(start_226)
ALI_226_2 = Date_ALI.index(end_226)

TEL_226_1 = Date_TEL.index(start_226)
TEL_226_2 = Date_TEL.index(end_226)

URC_226_1 = Date_URC.index(start_226)
URC_226_2 = Date_URC.index(end_226)

PGOLD_226_1 = Date_PGOLD.index(start_226)
PGOLD_226_2 = Date_PGOLD.index(end_226)

#save the shortened versions of closing prices
β226 = np.save('SMshortPrices.npy', ClosePrice_SM[SM_226_1:SM_226_2])
E226 = np.save('ALIshortPrices.npy', ClosePrice_ALI[ALI_226_1:ALI_226_2])
V226 = np.save('TELshortPrices.npy', ClosePrice_TEL[TEL_226_1:TEL_226_2])
ε226 = np.save('URCshortPrices.npy', ClosePrice_URC[URC_226_1:URC_226_2])
W226 = np.save('PGOLDshortPrices.npy', ClosePrice_PGOLD[PGOLD_226_1:PGOLD_226_2])

# import modules
import numpy as np
# import prices
β226 = np.load('SMshortPrices.npy', allow_pickle=True).tolist()
E226 = np.load('ALIshortPrices.npy', allow_pickle=True).tolist()
V226 = np.load('TELshortPrices.npy', allow_pickle=True).tolist()
ε226 = np.load('URCshortPrices.npy', allow_pickle=True).tolist()
W226 = np.load('PGOLDshortPrices.npy', allow_pickle=True).tolist()
```

```python
# extract spatio-temporal sequences
STC_226 = []
st_226 = []
n = 1
while n < len(β226):
    sc_226 = ''
    if β226[n] > β226[n-1]:
        sc_226 = sc_226 + 'β'
    if E226[n] > E226[n-1]:
        sc_226 = sc_226 + 'E'
    if V226[n] > V226[n-1]:
        sc_226 = sc_226 + 'V'
    if ε226[n] > ε226[n-1]:
        sc_226 = sc_226 + 'ε'
    if W226[n] > W226[n-1]:
        sc_226 = sc_226 + 'W'
    if len(sc_226) > 0:
        st_226.append(sc_226)
    else:
        STC_226.append(st_226)
        st_226 = []
    n = n + 1


# record the number of empty spatio-temporal sequences
nSTCempty_226 = STC_226.count([])
# keep only non-empty spatio-temporal sequences
STC_226 = [ st_226 for st_226 in STC_226 if len(st_226) > 0 ]

# record the length of the longest spatio-temporal sequence
lmin_226 = len(β226)
lmax_226 = 0
for st_226 in STC_226:
    if len(st_226) < lmin_226:
        lmin_226 = len(st_226)
    if len(st_226) > lmax_226:
        lmax_226 = len(st_226)
        
# restrict spatio-temporal sequences to length-5
STC5_226 = [ st_226 for st_226 in STC_226 if len(st_226) <= 5 ]

# unpack spatio-temporal sequences to temporal sequences
TC5_226 = []
for n in range(len(STC5_226)):
    tc_226 = [ u for u in STC5_226[n][0] ]
    for k in range(1, len(STC5_226[n])):
        tc_226 = [ u+v for u in tc_226 for v in STC5_226[n][k] ]
    TC5_226 = TC5_226 + tc_226
setTC5_226 = list(set(TC5_226))

# get empirical frequencies
fTC5_226 = [ TC5_226.count(setTC5_226[l]) for l in range(len(setTC5_226)) ]

# statistical testing against null model
np.random.seed(0)
fTC5null_226 = [ [] for l in range(len(setTC5_226)) ]
for s in range(226):
    STC5null_226 = [ list(st_226) for st_226 in STC5_226]
    for st_226 in STC5null_226:
        np.random.shuffle(st_226)
    TC5null_226= []
    for n in range(len(STC5null_226)):
        tc_226 = [ u for u in STC5null_226[n][0] ]
        for k in range(1, len(STC5null_226[n])):
            tc_226 = [ u+v for u in tc_226 for v in STC5null_226[n][k] ]
        TC5null_226= TC5null_226+ tc_226
    for l in range(len(setTC5_226)):
        if TC5null_226.count(setTC5_226[l]) > 0:
            fTC5null_226[l].append(TC5null_226.count(setTC5_226[l]))
        else:
            fTC5null_226[l].append(0)
```

```python
print("List of Statistically Significant Sequences at α = 0.05")

#statistical significance testing at p < 0.05 level
for l in range(len(setTC5_226)):
    if sum(freq >= fTC5_226[l] for freq in fTC5null_226[l]) < 5:
        print(setTC5_226[l], fTC5_226[l], sum(freq >= fTC5_226[l] for freq in fTC5null_226[l]))
```

```python
print("List of Statistically Significant Sequences at α = 0.01")

#statistical significance testing at p < 0.01 level
for l in range(len(setTC5_226)):
    if sum(freq >= fTC5_226[l] for freq in fTC5null_226[l]) < 1:
        print(setTC5_226[l], fTC5_226[l], sum(freq >= fTC5_226[l] for freq in fTC5null_226[l]))
```

```python
# extract spatio-temporal sequences
STC_237 = []
st_237 = []
n = 1
while n < len(Q237):
    sc_237 = ''
    if Q237[n] > Q237[n-1]:
        sc_237 = sc_237 + 'Q'
    if E237[n] > E237[n-1]:
        sc_237 = sc_237 + 'E'
    if Z237[n] > Z237[n-1]:
        sc_237 = sc_237 + 'Z'
    if ε237[n] > ε237[n-1]:
        sc_237 = sc_237 + 'ε'
    if W237[n] > W237[n-1]:
        sc_237 = sc_237 + 'W'
    if len(sc_237) > 0:
        st_237.append(sc_237)
    else:
        STC_237.append(st_237)
        st_237 = []
    n = n + 1


# record the number of empty spatio-temporal sequences
nSTCempty_237 = STC_237.count([])
# keep only non-empty spatio-temporal sequences
STC_237 = [ st_237 for st_237 in STC_237 if len(st_237) > 0 ]

# record the length of the longest spatio-temporal sequence
lmin_237 = len(Q237)
lmax_237 = 0
for st_237 in STC_237:
    if len(st_237) < lmin_237:
        lmin_237 = len(st_237)
    if len(st_237) > lmax_237:
        lmax_237 = len(st_237)
        
# restrict spatio-temporal sequences to length-5
STC5_237 = [ st_237 for st_237 in STC_237 if len(st_237) <= 5 ]

# unpack spatio-temporal sequences to temporal sequences
TC5_237 = []
for n in range(len(STC5_237)):
    tc_237 = [ u for u in STC5_237[n][0] ]
    for k in range(1, len(STC5_237[n])):
        tc_237 = [ u+v for u in tc_237 for v in STC5_237[n][k] ]
    TC5_237 = TC5_237 + tc_237
setTC5_237 = list(set(TC5_237))

# get empirical frequencies
fTC5_237 = [ TC5_237.count(setTC5_237[l]) for l in range(len(setTC5_237)) ]

# statistical testing against null model
np.random.seed(0)
fTC5null_237 = [ [] for l in range(len(setTC5_237)) ]
for s in range(237):
    STC5null_237 = [ list(st_237) for st_237 in STC5_237]
    for st_237 in STC5null_237:
        np.random.shuffle(st_237)
    TC5null_237= []
    for n in range(len(STC5null_237)):
        tc_237 = [ u for u in STC5null_237[n][0] ]
        for k in range(1, len(STC5null_237[n])):
            tc_237 = [ u+v for u in tc_237 for v in STC5null_237[n][k] ]
        TC5null_237= TC5null_237+ tc_237
    for l in range(len(setTC5_237)):
        if TC5null_237.count(setTC5_237[l]) > 0:
            fTC5null_237[l].append(TC5null_237.count(setTC5_237[l]))
        else:
            fTC5null_237[l].append(0)
```

```python
print("List of Statistically Significant Sequences at α = 0.05")

#statistical significance testing at p < 0.05 level
for l in range(len(setTC5_237)):
    if sum(freq >= fTC5_237[l] for freq in fTC5null_237[l]) < 5:
        print(setTC5_237[l], fTC5_237[l], sum(freq >= fTC5_237[l] for freq in fTC5null_237[l]))
```

```python
print("List of Statistically Significant Sequences at α = 0.01")

#statistical significance testing at p < 0.01 level
for l in range(len(setTC5_237)):
    if sum(freq >= fTC5_237[l] for freq in fTC5null_237[l]) < 1:
        print(setTC5_237[l], fTC5_237[l], sum(freq >= fTC5_237[l] for freq in fTC5null_237[l]))
```
