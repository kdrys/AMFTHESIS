```python
#shorten the closing prices
AP_46_1 = Date_AP.index(start_46)
AP_46_2 = Date_AP.index(end_46)

AC_46_1 = Date_AC.index(start_46)
AC_46_2 = Date_AC.index(end_46)

FGEN_46_1 = Date_FGEN.index(start_46)
FGEN_46_2 = Date_FGEN.index(end_46)

SMC_46_1 = Date_SMC.index(start_46)
SMC_46_2 = Date_SMC.index(end_46)

SCC_46_1 = Date_SCC.index(start_46)
SCC_46_2 = Date_SCC.index(end_46)
```

```python
#save the shortened versions of closing prices
B46 = np.save('APshortPrices.npy', ClosePrice_AP[AP_46_1:AP_46_2])
D46 = np.save('ACshortPrices.npy', ClosePrice_AC[AC_46_1:AC_46_2])
J46 = np.save('FGENshortPrices.npy', ClosePrice_FGEN[FGEN_46_1:FGEN_46_2])
Y46 = np.save('SMCshortPrices.npy', ClosePrice_SMC[SMC_46_1:SMC_46_2])
α46 = np.save('SCCshortPrices.npy', ClosePrice_SCC[SCC_46_1:SCC_46_2])
```

```python
# import modules
import numpy as np
# import prices

B46 = np.load('APshortPrices.npy', allow_pickle=True).tolist() 
D46 = np.load('ACshortPrices.npy', allow_pickle=True).tolist() 
J46 = np.load('FGENshortPrices.npy', allow_pickle=True).tolist() 
Y46 = np.load('SMCshortPrices.npy', allow_pickle=True).tolist() 
α46 = np.load('SCCshortPrices.npy', allow_pickle=True).tolist()
```

```python
# extract spatio-temporal sequences
STC_46 = []
st_46 = []
n = 1
while n < len(B46):
    sc_46 = ''
    if B46[n] > B46[n-1]:
        sc_46 = sc_46 + 'B'
    if D46[n] > D46[n-1]:
        sc_46 = sc_46 + 'D'
    if J46[n] > J46[n-1]:
        sc_46 = sc_46 + 'J'
    if Y46[n] > Y46[n-1]:
        sc_46 = sc_46 + 'Y'
    if α46[n] > α46[n-1]:
        sc_46 = sc_46 + 'α'
    if len(sc_46) > 0:
        st_46.append(sc_46)
    else:
        STC_46.append(st_46)
        st_46 = []
    n = n + 1
```

```python
# record the number of empty spatio-temporal sequences
nSTCempty_46 = STC_46.count([])
# keep only non-empty spatio-temporal sequences
STC_46 = [ st_46 for st_46 in STC_46 if len(st_46) > 0 ]

# record the length of the longest spatio-temporal sequence
lmin_46 = len(B46)
lmax_46 = 0
for st_46 in STC_46:
    if len(st_46) < lmin_46:
        lmin_46 = len(st_46)
    if len(st_46) > lmax_46:
        lmax_46 = len(st_46)
        
# restrict spatio-temporal sequences to length-5
STC5_46 = [ st_46 for st_46 in STC_46 if len(st_46) <= 5 ]

# unpack spatio-temporal sequences to temporal sequences
TC5_46 = []
for n in range(len(STC5_46)):
    tc_46 = [ u for u in STC5_46[n][0] ]
    for k in range(1, len(STC5_46[n])):
        tc_46 = [ u+v for u in tc_46 for v in STC5_46[n][k] ]
    TC5_46 = TC5_46 + tc_46
setTC5_46 = list(set(TC5_46))

# get empirical frequencies
fTC5_46 = [ TC5_46.count(setTC5_46[l]) for l in range(len(setTC5_46)) ]

# statistical testing against null model
np.random.seed(0)
fTC5null_46 = [ [] for l in range(len(setTC5_46)) ]
for s in range(100):
    STC5null_46 = [ list(st_46) for st_46 in STC5_46]
    for st_46 in STC5null_46:
        np.random.shuffle(st_46)
    TC5null_46= []
    for n in range(len(STC5null_46)):
        tc_46 = [ u for u in STC5null_46[n][0] ]
        for k in range(1, len(STC5null_46[n])):
            tc_46 = [ u+v for u in tc_46 for v in STC5null_46[n][k] ]
        TC5null_46= TC5null_46+ tc_46
    for l in range(len(setTC5_46)):
        if TC5null_46.count(setTC5_46[l]) > 0:
            fTC5null_46[l].append(TC5null_46.count(setTC5_46[l]))
        else:
            fTC5null_46[l].append(0)
```

```python
print("List of Statistically Significant Sequences at R = 0.05")

#statistical significance testing at p < 0.05 level
for l in range(len(setTC5_46)):
     if sum(freq >= fTC5_46[l] for freq in fTC5null_46[l]) < 5:
#         print(setTC5_46[l], fTC5_46[l], sum(freq >= fTC5_46[l] for freq in fTC5null_46[l]))
        print(setTC5_46[l])
```

```python
print("List of Statistically Significant Sequences at R = 0.01")

#statistical significance testing at p < 0.01 level
for l in range(len(setTC5_46)):
     if sum(freq >= fTC5_46[l] for freq in fTC5null_46[l]) < 1:
#         print(setTC5_46[l], fTC5_46[l], sum(freq >= fTC5_46[l] for freq in fTC5null_46[l]))
        print(setTC5_46[l])
```
